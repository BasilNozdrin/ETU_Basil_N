## SIR модель. Метод Рунге-Кутта 4 порядка

**S** -- susceptible individuals $\quad$ **I** -- infected individuals $\quad$ **R** -- recovered individuals  
**N** -- population size $\quad$ **betta** -- заразность ($\beta$) $\quad$ **gamma** -- смертность ($\gamma$)  
$$k_1^S = \frac{-\beta s i }{N}, \quad k_1^I = \frac{\beta s i }{N}-ki, \quad k_1^R = ki$$
https://blog.tonytsai.name/blog/2014-11-24-rk4-method-for-solving-sir-model/

```python
S = [1000]
I = [300]
R = [0]
N = S[0]+I[0]+R[0]

betta = 0.5
gamma = 0.33
h = 1
n = 100

for j in range(1, n):
    s = S[j-1]
    i = I[j-1]
    r = R[j-1]
    
    # вывод коэфициентов k1
    k1_S = (-betta * s * i)/N
    k1_I = (betta * s * i)/N - gamma*i
    k1_R = gamma*i
    # вывод коэфициентов k2
    k2_S = -(betta/N)*(s+k1_S*h/2)*(i+k1_I*h/2)
    k2_I = (betta/N)*(s+k1_S*h/2)*(i+k1_I*h/2)-gamma*(i+k1_I*h/2)
    k2_R = gamma*(i+k1_I*h/2)
    # вывод коэфициентов k3
    k3_S = -(betta/N)*(s+k2_S*h/2)*(i+k2_I*h/2)
    k3_I = (betta/N)*(s+k2_S*h/2)*(i+k2_I*h/2)-gamma*(i+k2_I*h/2)
    k3_R = gamma*(i+k2_I*h/2)
    # вывод коэфициентов k4
    k4_S = -(betta/N)*(s+k3_S*h)*(i+k3_I*h)
    k4_I = (betta/N)*(s+k3_S*h)*(i+k3_I*h)-gamma*(i+k3_I*h)
    k4_R = gamma*(i+k3_I*h)
    
    # основная формула
    S_i = s+(h/6)*(k1_S + 2*k2_S + 2*k3_S + k4_S)
    I_i = i+(h/6)*(k1_I + 2*k2_I + 2*k3_I + k4_I)
#     R_i = r+(h/6)*(k1_R+2*k2_R+2*k3_R+k4_R)
    R_i = N - S_i - I_i
    S.append(S_i)
    I.append(I_i)
    R.append(R_i)


artist = plt.figure()
ax = artist.add_axes([0,0,1,1])
ax.set_xlabel("days")
ax.set_ylabel("individuals")
ax.plot(range(n), S, label = 'S')
ax.plot(range(n), I, label = 'I')
ax.plot(range(n), R, label = 'R')
legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.show()
```

## Модель хищник-жертва. Метод Рунге-Кутта 4 порядка  
  
**R_n** - число жертв $\quad\quad\quad\quad\quad\quad\quad$ **F_n** - число хищников  
**s** - коэф. смертности хищников $\quad\quad$ **g** - коэффициент рождаемости хищников  
**a** - скорость размножения жертв $\quad\quad$ **b** - коэффициент смертности жертв  
**n** - количество шагов $\quad\quad\quad\quad\quad\quad$ **h** - размер шага  
https://iopscience.iop.org/article/10.1088/1742-6596/1341/6/062024/pdf

```python
R = [10]
F = [2]
g = 0.5 
s = 2 
a = 1
b = 0.5 
h = 0.01 
n = 2000 


def f1(x, y):
    return a*x - b*x*y


def f2(x, y):
    return g*x*y - s*y


for i in range(1, n):
    k_1_r = f2(F[i-1], R[i-1])
    k_1_f = f1(F[i-1], R[i-1])
    
    k_2_r = f2(F[i-1]+h/2*k_1_f, R[i-1]+h/2*k_1_r)
    k_2_f = f1(F[i-1]+h/2*k_1_f, R[i-1]+h/2*k_1_r)
    
    k_3_r = f2(F[i-1]+h/2*k_2_f, R[i-1]+h/2*k_2_r)
    k_3_f = f1(F[i-1]+h/2*k_2_f, R[i-1]+h/2*k_2_r)
    
    k_4_r = f2(F[i-1]+h*k_3_f, R[i-1]+h*k_3_r)
    k_4_f = f1(F[i-1]+h*k_3_f, R[i-1]+h*k_3_r)
    
    R_i = R[i-1]+(h/6)*(k_1_r + 2*k_2_r + 2*k_3_r + k_4_r)
    F_i = F[i-1]+(h/6)*(k_1_f + 2*k_2_f + 2*k_3_f + k_4_f)
    
    R.append(R_i)
    F.append(F_i)

artist = plt.figure()
ax = artist.add_axes([0,0,1,1])
ax.set_xlabel("days")
ax.set_ylabel("creatures")
ax.plot(range(n), R, label = 'Хищники')
ax.plot(range(n), F, label = 'Жертвы')
legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.show()
```
