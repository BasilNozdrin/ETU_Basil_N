### Моделирование задачи **"хищник-жертва"** 

**F_n** -- хищники  
**R_n** -- жертвы  
**b(u)** -- скорость размножения  
**a** - процент удачи охот

```python
import matplotlib
import numpy
import math
```

```python
def genData(m, F, h, n):
    F_data = [F(0)]
    v_data = [0]
    x_data = [0]
    for i in range(1, n):
        F_data.append(max(0, int(F(i))))
        v_data.append(max(0, int(
            v_data[i-1] + h*F_data[i-1]/m)))
        x_data.append(max(0, int(
            x_data[i-1] + h*v_data[i-1])))
    return (F_data, v_data, x_data)
```

F(t) = const

```python
m = 1
def F(t):
    return 1

h = 1
n = 10
```

```python
%matplotlib inline

x = range(n)
y1, y2, y3 = genData(m, F, h, n)

artist = matplotlib.pyplot.figure()
ax = artist.add_axes([0,0,1,1])
# ax.set_xlabel("days")
# ax.set_ylabel("number of creatures")

ax.plot(x, y1, label='Force')
ax.plot(x, y2, label='Velocity')
ax.plot(x, y3, label='x coordinate')

legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')

# legend.get_frame().set_facecolor('C0')

matplotlib.pyplot.show()
```
