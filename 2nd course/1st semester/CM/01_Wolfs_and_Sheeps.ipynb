### Моделирование задачи **"хищник-жертва"** 

**F_n** -- хищники  
**R_n** -- жертвы  
**b(u)** -- скорость размножения  
**a** - процент удачи охот

```python
import matplotlib
import numpy
import math
```

```python
def genData(a, b, d, f, r, n):
    f = f[::]
    r = r[::]
    for i in range(1, n):
        f.append(max(0, f[i-1] - d*f[i-1] + a*f[i-1]*r[i-1]))
        r.append(max(0, r[i-1] + b(i)*r[i-1] - a*f[i-1]*r[i-1]))
    return (f, r)
```

первый набор данных

```python
a = 0.007
def b(n):
    return 0.7
d = 0.5
f = [5]
r = [200]
n = 20
```

```python
%matplotlib inline

x = range(n)
y1, y2 = genData(a, b, d, f, r, n)

artist = matplotlib.pyplot.figure()
ax = artist.add_axes([0,0,1,1])
ax.set_xlabel("days")
ax.set_ylabel("number of creatures")

ax.plot(x, y1, label='F')
ax.plot(x, y2, label='R')

legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')

# legend.get_frame().set_facecolor('C0')

matplotlib.pyplot.show()
```

второй набор данных

```python
a = 0.0005
def b(n):
    alpha = 0.005
    return math.exp(-alpha*n)
d = 0.5
f = [5]
r = [200]
n = 100
```

```python
%matplotlib inline

x = range(n)
y1, y2 = genData(a, b, d, f, r, n)

artist = matplotlib.pyplot.figure()
ax = artist.add_axes([0,0,1,1])
ax.set_xlabel("days")
ax.set_ylabel("number of creatures")

ax.plot(x, y1, label='F')
ax.plot(x, y2, label='R')

legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')

# legend.get_frame().set_facecolor('C0')

matplotlib.pyplot.show()
```

третий набор данных

```python
a = 0.007
b = [1000] 
d = 0.5
u = 100
f = [5]
r = [200]
n = 20
```

```python
def genData2(a, b, d, f, r, n):
    b = b[::]
    f = f[::]
    r = r[::]
    for i in range(1, n):
        b.append(max(0, b[i-1] - r[i-1]*0.3 + b[i-1]*0.7))
        f.append(max(0, f[i-1] - d*f[i-1] + a*f[i-1]*r[i-1]))
        r.append(max(0, r[i-1] + b[i-1]*r[i-1] - a*f[i-1]*r[i-1]))
    return (f, r)
```

```python
%matplotlib inline

x = range(n)
y1, y2 = genData2(a, b, d, f, r, n)

artist = matplotlib.pyplot.figure()
ax = artist.add_axes([0,0,1,1])
ax.set_xlabel("days")
ax.set_ylabel("number of creatures")

ax.plot(x, y1, label='F')
ax.plot(x, y2, label='R')

legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')

# legend.get_frame().set_facecolor('C0')

matplotlib.pyplot.show()
```
