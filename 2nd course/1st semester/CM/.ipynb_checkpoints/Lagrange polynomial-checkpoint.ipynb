```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Функция cos(x)/x и ее интерполяция многочленом Лагранджа

```python
def genuine(x):
    return math.cos(x)/x


def Lagrange(x,y,t):
    res = 0
    for j in range(len(y)):
        p1 = 1
        p2 = 1
        for i in range(len(x)):
            if i != j:
                p1 = p1*(t-x[i])
                p2 = p2*(x[j]-x[i])
        res += y[j]*p1/p2
    return res
```

```python
# f: [1, n] -> R
n = 30

# точность графиков
p1 = 300
p2 = 30


# data for cosx/x
X1 = np.linspace(1, n, p1)
Y1 = [genuine(i) for i in X1]

# interpolation
X2 = np.linspace(1, n, p2)
pts = [genuine(x) for x in X2]
Y2 = [Lagrange(X2, pts, x) for x in X2]

artist = plt.figure()
ax = artist.add_axes([0,0,1,1])
ax.plot(X1, Y1, 'm', color='blue',label='f(x)=cos(x)/x')
ax.plot(X2, Y2, 'ro', color='red', markersize=5, label='g(x)=L_f(x)')
legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.grid(True)
plt.show()
```
