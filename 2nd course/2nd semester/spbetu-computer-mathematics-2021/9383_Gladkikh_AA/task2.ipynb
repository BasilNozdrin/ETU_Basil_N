## Основные теоретические положения

### Орбита элемента

Для начала введем вспомогательное определение.

***Def 1*** Пусть ${G}$ - группа с нейтральным элементом ${e}$, ${X}$ - некое множество. Тогда говорят, что группа ${G}$ действует на множество ${X}$, если задана операция ${G \times X \rightarrow G}$. 
Данная операция обладает следующими свойствами ${\forall x \in X, \forall g, h \in G}$:
1. ${ex = x}$.
2. ${g(hx) = (gh)x}$.

***Def 2*** Орбитой элемента ${x \in X}$ под действием группы ${G}$ называется множество ${Gx = \left\{gx \vert g \in G\right\}}$.

***Def 3*** Длина орбиты - количество элементов в данной орбите.

### Стабилизатор элемента

***Def 4*** Неподвижными точками элемента ${g \in G}$ называют те элементы ${x \in X}$, для которых выполняется следующее ${gx = x}$.

Множество неподвижных точек элемента ${g}$ будем обозначать ${X^g}$.

***Def 5*** Множество элементов группы ${G}$, оставляющих на месте данный элемент ${x \in X}$, называется *стабилизатором* элемента ${x}$ и обозначается ${St(x) =  \left \{ g \in G \vert gx = x \right\} }$. Очевидно, что стабилизатор является подгруппой в ${G}$.

### Лемма Бёрнсайда 

Количество орбит действия группы ${G}$ на множество ${X}$ равно:

${N = \frac{1}{\left | G \right |}\sum_{g\in G}(\left| X^g \right|)}$

### Раскраски

Пусть ${I}$ - произвольное множество, ${C}$ - множество цветов.

***Def 6*** Раскраской множества ${I}$ называется некая функция ${f: I\rightarrow C}$.

Множество всех раскрасок будем обозначать ${C^I}$, а количество всех раскрасов ${{\left|C\right|}^{\left|I\right|}}$.

***Def 7*** Пусть ${G}$ - группа, действующая на множество ${I}$. Тогда формула ${gf(i)=f(g^{-1}i)}$ задает действие ${g}$ на ${C^I}$, где ${g\in G,i \in I, f\in C^I}$.

***Th*** Пусть у нас есть ${g \in G, f \in C^I}$. Для этого элемента ${g}$ и функции ${f}$ эквивалентны следующие условия:
1. ${gf = g}$.
2. ${f(i) = f(g^ni), \forall i \in I, \forall n\in Z}$.
3. Если элементы ${i,j \in I}$ лежат в одном цикле в циклической записи перестановки ${\alpha(g)}$, то ${f(i) = f(g)}$.

***Def 8*** Количество раскрасок из ${C^I}$, которые сохраняют данный элемент ${g \in G}$ равно ${\left|C\right|^{c(\alpha(g))}}$, где ${c(\alpha)}$ - количество независимых циклов в перестановке ${\alpha}$. 
- **NB!** также учитываются циклы длины 1.

### Поиск количества раскрасок

В нашей задаче на множество всех раскрасок ${X}$ будет действовать группа вращений ${G}$. Раскраски граней, совмещенные друг с другом каким-нибудь преобразованием из группы ${G}$, задают одинаковые тетраэдры, то есть лежат в одной орбите под действием группы ${G}$. Таким образом, для поиска числа раскрасок достаточно подсчитать количество орбит действия группы ${G}$ на множество ${X}$.

Воспользуемся леммой Бёрнсайда, благодаря которой узнаем, какое число раскрасок их ${X}$ оставляет на месте каждый элемент группы ${G}$.

Нейтральный элемент ${e = id}$ оставляет на своем месте все раскраски, значит мощность ${\left |X^e \right | = \left | X \right| = 4^4}$.

Согласно ***Def 8*** количество раскрасок, которые сохраняют данный элемент равно ${\left|C\right|^{c(\alpha(g))}}$, где ${c(\alpha)}$ - количество независимых циклов в перестановке ${\alpha}$. Тогда очевидно:

${\left |X^{\tau_1} \right | = 4^2}$

${\left |X^{\tau_1^2} \right | = 4^2}$

${\left |X^{\tau_2} \right | = 4^2}$

${\left |X^{\tau_2^2} \right | = 4^2}$

${\left |X^{\tau_3} \right | = 4^2}$

${\left |X^{\tau_3^2} \right | = 4^2}$

${\left |X^{\tau_4} \right | = 4^2}$

${\left |X^{\tau_4^2} \right | = 4^2}$

Аналогично несложно заметить, что:

${\left |X^{\sigma_1} \right | = 4^2}$

${\left |X^{\sigma_2} \right | = 4^2}$

${\left |X^{\sigma_3} \right | = 4^2}$

Порядок группы ${G}$ равен ${8 + 3 + 1 = 12}$ - 11 вращений и тождественное преобразование. Тогда по лемме Бёрнсайда:

${N = \frac{1}{12}\left(4^4 + 11 \cdot 4^2\right) = 36}$

#### Ответ: 36.

## 2. Программная реализация

```python
import sage.all
from sage.plot.plot3d.index_face_set import IndexFaceSet
from sage.plot.plot3d.texture import Texture
from sage.plot.plot3d.shapes import Text
from sage.plot.plot3d.shapes2 import text3d

import numpy as np

import copy
import random
```

```python
class Tetrahedron:
    def __init__(self, coords=(0, 0, 0), color="blue", opacity=0.1):
        self.coords = coords
        self.opacity = opacity
        self.t = tetrahedron(self.coords, color=color, opacity=self.opacity)
        self.face_amount = 4
        self.colors = {0: color, 1: color, 2: color, 3: color}
        self.textColor = "green"
        
    def calculateTextPosition(self):
        textPos = []
        for face in self.t.faces():
            AB_vec = np.array([face[1][0] - face[0][0], face[1][1] - face[0][1], face[1][2] - face[0][2]])
            AC_vec = np.array([face[2][0] - face[0][0], face[2][1] - face[0][1], face[2][2] - face[0][2]])
            
            normal_vec = np.cross(AB_vec, AC_vec)
            normal_vec /= np.sum(np.square(normal_vec))
            
            x_mid = (face[0][0] + face[1][0] + face[2][0]) / len(face) + normal_vec[0]
            y_mid = (face[0][1] + face[1][1] + face[2][1]) / len(face) + normal_vec[1]
            z_mid = (face[0][2] + face[1][2] + face[2][2]) / len(face) + normal_vec[2]
            textPos.append((x_mid, y_mid, z_mid))
        return textPos
    
    def changeColor(self, newColorMap):
        for key in range(self.face_amount):
            if newColorMap.get(key) != None:
                self.colors[key] = newColorMap[key]
                
    def changeTextColor(self, newTextColor):
        self.textColor = newTextColor
        
    def rotate(self, permutation):
        newColors = {}
        for key in permutation:
            newColors[permutation[key] - 1] = self.colors[key - 1]
        self.colors = newColors
        
    def printColors(self):
        for key in range(self.face_amount):
            if key + 1 != self.face_amount:
                print(key + 1, ":", self.colors[key], end=', ')
            else:
                print(key + 1, ":", self.colors[key], end='')
        print('\n')
        
    def show(self, showText=true):
        self.printColors()
        textures = []
        for key in range(self.face_amount):
            color = self.colors[key]
            textures.append(Texture(color))
        obj = IndexFaceSet(self.t.faces(), texture_list=textures)
        
        if showText:
            textPos = self.calculateTextPosition()
            for ind, pos in enumerate(textPos):
                text = text3d(str(ind + 1), pos, fontsize=100, fontweight=800, color=self.textColor)
                obj += text
        obj.show()
```

```python
def tau1(tetrahedron):
    permutation = {1: 1, 2: 4, 3: 2, 4: 3}
    tetrahedron.rotate(permutation)

def tau2(tetrahedron):
    permutation = {1: 3, 2: 2, 3: 4, 4: 1}
    tetrahedron.rotate(permutation)

def tau3(tetrahedron):
    permutation = {1: 2, 2: 4, 3: 3, 4: 1}
    tetrahedron.rotate(permutation)
    
def tau4(tetrahedron):
    permutation = {1: 3, 2: 1, 3: 2, 4: 4}
    tetrahedron.rotate(permutation)

def sigma1(tetrahedron):
    permutation = {1: 3, 2: 4, 3: 1, 4: 2}
    tetrahedron.rotate(permutation)
    
def sigma2(tetrahedron):
    permutation = {1: 4, 2: 3, 3: 2, 4: 1}
    tetrahedron.rotate(permutation)

def sigma3(tetrahedron):
    permutation = {1: 2, 2: 1, 3: 4, 4: 3}
    tetrahedron.rotate(permutation)
```

```python
def randomColors():
    colors = ["white", "red", "orange", "cyan"]
    random.shuffle(colors)
    colorsDict = {}
    for i in range(4):
        colorsDict[i] = colors[i]
    return colorsDict

myTetr = Tetrahedron()

tetrColors = randomColors()

myTetr.changeColor(tetrColors)

myTetr.show()
```

## 3. Общее решение

Количество независимых циклов в перестановках не зависит от числа цветов в раскраске. Поэтому если в раскраске ${n}$ цветов, то:

${\left |X^e \right | = n^4}$

${\left |X^{\tau_1} \right | = n^2}$

${\left |X^{\tau_1^2} \right | = n^2}$

${\left |X^{\tau_2} \right | = n^2}$

${\left |X^{\tau_2^2} \right | = n^2}$

${\left |X^{\tau_3} \right | = n^2}$

${\left |X^{\tau_3^2} \right | = n^2}$

${\left |X^{\tau_4} \right | = n^2}$

${\left |X^{\tau_4^2} \right | = n^2}$

Аналогично:

${\left |X^{\sigma_1} \right | = n^2}$

${\left |X^{\sigma_2} \right | = n^2}$

${\left |X^{\sigma_3} \right | = n^2}$

Порядок группы ${G}$ также не зависит от числа цветов и равен ${8 + 3 + 1 = 12}$. Тогда по лемме Бёрнсайда:

${N = \frac{1}{12}\left(n^4 + 11 \cdot n^2\right)}$
