## Основные теоретические положения
***Def. 1*** пусть $G$ - группа, $e$ -  нейтральный элемент группы $G$, $X$ - множество. Тогда $G$ действует на $X$, если задана операция $G x X -> G$. Тогда $\forall x \in X$,  $\forall g,h \in G$: 1)$ex=x$, 2) $g(hx)=(gh)x$  

***Def. 2*** *орбитой* элемента $ x \in X$ под действием $G$ называется множество $Gx=\{gx|g \in G\}$. Количество элементов в этом множестве называется длиной орбиты.  

***Def. 3*** *неподвижными точками* $ g \in G: x \in X: gx = x$. $X^g$ - множество неподвижных точек элемента $g$.  

***Лемма Бёрнсайда***  
Количество орбит действия группы $G$ на множество $X$ вычисляется по формуле: $N= \frac{1}{_{\#G}} \cdot \sum_{g \in G}\# X^g $

***Def. 4*** пусть $I$ - произвольное множество, $C$ - множество цветов. *Раскраской* множетсва $I$ называется функция $I->G$. $C^I$ - множество всех раскрасок. $\# C^I=(\#C)^{\#I}$

***Def. 5*** пусть $G$ - группа, действует на множестве $I$, тогда $gf(i)=f(g^{-1}i)$, где $g \in G, i \in I, f \in C^I$ задаёт действие $G$ на множество $C^I$

***Лемма 2***  
Для элемента $g \in G$ и функци $f \in C^I$ следующие условия эквивалентны:
* $gf = f$
* $f(i)=f(g^ni), \forall i \in I, n \in Z$
* если $i, j \in I$ лежат в одном цикле в циклической записи в перестановке $\alpha(g)$, то $f(i)=f(j)$

***Def. 6*** Количество раскрасок $C^I$, которые сохраняют $g \in G$ равно $(\# C)^{C(\alpha(g))}$, где $C$ - количество независимых циклов в записи перестановки

На множестве введено тождественное преобразование, запишем его в виде перестановки:
$e = \begin{pmatrix}
1 & 2 & ... & 51 \\\
1 & 2 & ... & 51
\end{pmatrix} = (1)(2)...(51)$

$\tau$ - поворот, переводящий каждую бусинку в следующую  
$\sigma$ - осевая симметрия  

Так как на данном множестве действует количестdj перестановок большое для того, чтобы выписывать все из них, выпишем несколько из них

$\tau = \begin{pmatrix}
1 & 2 & ... & 51 \\\
2 & 3 & ... & 1
\end{pmatrix} = (1 \ 2 ... 51)$

$\tau^2 = \begin{pmatrix}
1 & 2 & ... & 51 \\\
3 & 4 & ... & 2
\end{pmatrix} = (1 \ 3 ... 49 \ 51)(2 \ 4 ...  48 \ 50)$

...

$\sigma = \begin{pmatrix}
1 & 2 & ...& 50 & 51 \\\
51 & 50 & ... & 2 & 1
\end{pmatrix} = (1 \ 51)(2 \ 50)...(25 \ 27)(26)$

$\sigma^2 = e$ 

Перестановка $\tau$ - цикл длины 51. По лемме 2 посчитаем количество циклов в перестановках $\tau^i \forall i$. Разложим $51$ на простые множители: $51 = 17*3$. $\tau^i$ является произведением из $d$ независимых циклов, где $d$=НОД($51, i$). Тогда для всех $i$ не кратных $17$ или $3$ $d$ будет иметь значение $1$.

Перестановка $\sigma$ - произведение $26$ независимых циклов. По определению $6$ это количество элементов из множества $X$, которые оставляют на месте каждый элемент группы $G$. После применения к перестановке $\sigma \ \tau^i$ результатом будет также произведение $26$-ти независимых циклов. Это происходит в силу симметрии структуры перестановки.  

Раскраски будут задавать одинаковые бусы тогда и только тогда, когда они лежат в одной орбите под действием $G$. Нужно вычислить количество орбит действия $G$ на $X$.

$|X^e| \ = \ 5^{51}$  
$|X^{\tau^i}| = 5^1$, где $i \in \{1..50\},$ исключая $i\%17 = 0$ и $i\%3 = 0$  
$|X^{\tau^{3}}| = |X^{\tau^{6}}| = |X^{\tau^{9}}| = ... = |X^{\tau^{48}}| = 5^{3}$  
$|X^{\tau^{17}}| = |X^{\tau^{34}}| = 5^{17}$   
$|X^{\sigma}| = |X^{\sigma \tau}| = |X^{\sigma \tau^2}| = ... = |X^{\sigma \tau^{50}}| = 5^{26}$  
$|G| = 102$

Тогда по лемме Бёрнсайда посчитаем количество различных бус:  
кол-во различных бус = $\frac{1}{102}(5^{51} + 32*5^{1} + 16*5^{3} + 2*5^{17} + 51*5^{26})$
