## Основные теоретические положения

Пусть $ G $ &mdash; группа с нейтральным элементом $ e $, $ X $ &mdash; множество. Тогда $ G $ называется действием на множество $ X $, если задана операция $ G \times X \to G $, обладающая следующими свойствами:

1. $ \forall x \in X: ex = x $
2. $ \forall x \in X, g, h \in G: g(hx) = (gh)x $

Орбитой элемента $ x \in X $ под действием группы $ G $ называется множество $ Gx = \left\{ gx \; \middle| \; g \in G \right\} $.

Количество элементов множества $ Gx $ называется длиной орбиты.

Неподвижными точками элемента $ g \in G $ называются такие $ x \in X $, для которых $ gx = x $.

Множество неподвижных точек элемента $ g \in G $ обозначается $ X^{g} $.

**Лемма Бёрнсайда.** Количество орбит действия группы $ G $ на множество $ X $ равно:

<center> $ N = \dfrac{1}{\#G} \displaystyle\sum\limits_{g \in G} \#X^g $ </center>

*Доказательство:* Сопоставим каждому элементу $ x \in X $ число $ \dfrac{1}{\# Gx} $, следовательно, $ \displaystyle\sum\limits_{x \in Gx} \dfrac{1}{\# Gx} = 1 $. Поэтому 

<center> $ N = \displaystyle\sum\limits_{x \in X} \dfrac{1}{\# Gx} = \displaystyle\sum\limits_{g \in G} \dfrac{\# X^g}{\# G} = \dfrac{1}{\# G} \displaystyle\sum\limits_{g \in G} \# X^g $, </center>

что и требовалось доказать.

**Постановка задачи о раскрасках.** Пусть $ I $ &mdash; произвольное множество, $ C $ &mdash; множество цветов. Тогда раскраской множества $ I $ называется функция $ I \to C $.

Множество всех раскрасок обозначается $ C^I $.

<center> $ \# C^I = (\# C)^{\# I} $ </center>

Пусть группа $ G $ действует на множество $ I $, тогда формула

<center> $ gf \left( i \right) = f \left( g^{-1} \left( i \right) \right), \qquad g \in G, \; i \in I, \; f \in C^I $ </center>

задаёт действие $ G $ на множество $ C^I $.

Для $ g \in G $ и $ f \in C^I $ эквивалентны следующие условия:

1. $ gf = f $
2. $ \forall i \in I, n \in \mathbb{Z}: f \left( i \right) = f \left( g^n i \right) $
3. Если $ i, j \in I $ лежат в одном цикле в циклической записи перестановки $ \varphi \left( g \right) $, то $ f \left( i \right) = f \left( j \right) $

Количество раскрасок из множества $ C^I $, которые сохраняют элемент $ g \in G $ равно $ \left( \# C \right)^{c \left( \varphi \left( g \right) \right)} $, где $ c \left( \varphi \left( g \right) \right) $ &mdash; количество независимых циклов в циклической записи перестановки $ \varphi \left( g \right) $.

*Доказательство:* Пусть $ \left< g \right> $ &mdash; циклическая подгруппа, порождённая элементом $ g $, $ \left< g \right> \! i $ &mdash; орбита элемента $ i $, $ Q $ &mdash; множество различных орбит. Так как $ gf = f $, то все элементы одной орбиты отображаются в одно и то же под действием функции $ f $, поэтому $ f $ однозначно определяется функцией $ \tilde{f} : Q \to C $, где $ f \left( \left< g \right> \! i \right) = f \left( i \right) $, то есть множество $ \left( C^I \right)^{g} $ биективно отображается на множество $ C^Q $, следовательно, $ \# \left( C^I \right)^{g} = \# C^Q $. Количество всех функций из $ Q $ в $ C $ равно $ \left( \# C \right)^{\# Q} = \left( \# C \right)^{c \left( \varphi \left( g \right) \right)} $, что и требовалось доказать.

## Выполнение работы

**Задача.** Круг разбит на 45 секторов, каждый из которых покрашен в один из 5 цветов. Сколькими способами можно составить такую мозаику (с точностью до поворота круга)?

Пусть $ X $ &mdash; множество секторов круга, $ C $ &mdash; множество цветов. Для данной задачи $ X = \left\{ x_{1}, x_{2}, \dots, x_{45} \right\} $, $ C = \left\{ a, b, c, d, e \right\} $.

Пусть $ G $ &mdash; группа преобразований множества $ X $, $ e $ &mdash; тождественное преобразование, $ \tau $ &mdash; поворот круга на $ 8^{\circ} $. Тогда:

$ \quad e = \begin{pmatrix}
    x_{1} & x_{2} & x_{3} & \cdots & x_{45} \\
    x_{1} & x_{2} & x_{3} & \cdots & x_{45} 
\end{pmatrix} = \begin{pmatrix} x_{1} \end{pmatrix} \begin{pmatrix} x_{2} \end{pmatrix} \begin{pmatrix} x_{3} \end{pmatrix} \dots \begin{pmatrix} x_{45} \end{pmatrix} $

$ \quad \tau = \begin{pmatrix}
    x_{1} & x_{2} & x_{3} & \cdots & x_{45} \\
    x_{2} & x_{3} & x_{4} & \cdots & x_{1} 
\end{pmatrix} = \begin{pmatrix} x_{1} & x_{2} & x_{3} & \cdots & x_{45} \end{pmatrix} $

$ \quad \tau^{2} = \begin{pmatrix}
    x_{1} & x_{2} & x_{3} & \cdots & x_{45} \\
    x_{3} & x_{4} & x_{5} & \cdots & x_{2} 
\end{pmatrix} = \begin{pmatrix} x_{1} & x_{3} & x_{5} & \cdots & x_{45} & x_{2} & x_{4} & \cdots & x_{44} \end{pmatrix} $

$ \quad \tau^{3} = \begin{pmatrix}
    x_{1} & x_{2} & x_{3} & \cdots & x_{45} \\
    x_{4} & x_{5} & x_{6} & \cdots & x_{3} 
\end{pmatrix} = \begin{pmatrix} x_{1} & x_{4} & x_{7} & \cdots & x_{43} \end{pmatrix} \begin{pmatrix} x_{2} & x_{5} & x_{8} & \cdots & x_{44} \end{pmatrix} \begin{pmatrix} x_{3} & x_{6} & x_{9} & \cdots & x_{45} \end{pmatrix} $

$ \qquad \cdots $

$ \quad \tau^{44} = \begin{pmatrix}
    x_{1} & x_{2} & x_{3} & \cdots & x_{45} \\
    x_{45} & x_{1} & x_{2} & \cdots & x_{44} 
\end{pmatrix} = \begin{pmatrix} x_{45} & x_{44} & x_{43} & \cdots & x_{1} \end{pmatrix} $

$ \quad \tau^{45} = \begin{pmatrix}
    x_{1} & x_{2} & x_{3} & \cdots & x_{45} \\
    x_{1} & x_{2} & x_{3} & \cdots & x_{45} 
\end{pmatrix} = e $

$ G = \left\{ e, \tau, \tau^{2}, \dots, \tau^{44} \right\} \Rightarrow \# G = 45 $.

Определим количество независимых циклов во всех перестановках $ \tau^{k} $, где $ k \in \left\{ 1, 2, \dots, 45 \right\} $. Поскольку $ 45 = 3^{2} \cdot 5 $, то рассмотрим все $ k $, кратные $ 3 $, $ 5 $, $ 9 $ и $ 15 $. Тогда по лемме Бёрнсайда число всевозможных раскрасок с точностью до поворота круга равно:

<center> $ N = \dfrac{1}{45} \left( 5^{45} + 2 \cdot 5^{15} + 4 \cdot 5^{9} + 6 \cdot 5^{5} + 8 \cdot 5^{3} + 24 \cdot 5 \right) $. </center>

Для решения задачи в общем виде была реализована функция getCircle():

```python
from math import sqrt
import matplotlib.pyplot as plt
from random import getrandbits


def getCircle(n, c, num):
    size = [1 for i in range(n)]
    colors = []
    num = num % (c ** n)
    clr = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#FF00FF', '#00FFFF', '#F0F0F0', '#FFFFFF']
    if c < 9:
        clr = clr[:c]
    else:
        for i in range(c - 8):
            flag = False
            while flag == False:
                cl = str(hex(getrandbits(24)))[2:]
                zeros = ['0' for j in range(6 - len(cl))]
                cl = '#' + ''.join(zeros) + cl
                if cl not in clr:
                    clr.append(cl)
                    flag = True
    for i in range(n):
        colors.append(clr[num % c])
        num = num // c
    plt.figure(figsize = (30, 30))
    plt.pie(size, colors = colors, wedgeprops = {'linewidth': 3, 'edgecolor': '#FFFFFF'})
    plt.show()
    dividers = {}
    viewed = []
    for i in range(n // 2, 1, -1):
        if n % i == 0:
            multiples = []
            for j in range(i, n, i):
                if j not in viewed:
                    viewed.append(j)
                    multiples.append(j)
            dividers.update({i: multiples})
    sum = ['{}^{{{}}}'.format(c, n)]
    remainder = n - 1
    for k in dividers:
        degree = len(dividers[k])
        sum.append('{} \\cdot {}^{{{}}}'.format(degree, c, k))
        remainder -= degree
    sum.append('{} \\cdot {}'.format(remainder, c))
    return html('$ \\quad N = \\dfrac{{1}}{{{}}} \\left( {} \\right) $'.format(n, ' + '.join(sum)))
```

Отобразим графически раскраску для поставленной задачи ($ \# X = 45 $, $ \# C = 5 $):

```python
show(getCircle(45, 5, 53 ** 19))
```

Отобразим другую раскраску для той же задачи:

```python
show(getCircle(45, 5, 19 ** 53))
```

Отобразим графически раскраску и решение для задачи при $ \# X = 51 $, $ \# C = 7 $:

```python
show(getCircle(51, 7, 45 ** 45))
```

Отобразим графически раскраску и решение для задачи при $ \# X = 60 $, $ \# C = 12 $:
