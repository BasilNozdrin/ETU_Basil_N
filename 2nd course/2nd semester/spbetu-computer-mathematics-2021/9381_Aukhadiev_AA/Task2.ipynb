## Основные теоретические положения

Def 1.1
Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая для $\forall x \in X$ и $\forall g,h \in G$ следующими свойствами:

1) $ex = x$

2) $g(hx) = (gh)x$

где $gx$ - это образ пары $(g,x)$.

Def 1.2
Пусть $X$ - множество. Множество всех биективных функций с $X \to X$ с операцией композиции называется симметрической группой на множестве $X$ и обозначается как $S_{X}$

Def 1.3
Орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в данной орбите называется длиной орбиты. Любые две орбиты либо не пересекаются, либо совпадают. Таким образом, множество $X$ разбивается в дизъюнктивное объединение орбит.

Def 1.4
Неподвижными точками элемента $g \in G$ называются те $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента g обозначается через $X^{g}$.

Th 1 (Burnside)
Количество орбит действия группы $G$ на множестве $X$ равно:

$\frac{1}{|G|}\sum_{g \in G} |X^{g}|$

Lemma 1
Предположим, что $\sigma$ является циклом длины $k$, а $НОД(k,n) = d$. Тогда перестановка $\sigma^{n}$ является произведением d циклов длины $\frac{k}{d}$.

Def 1.5
Пусть $I$ - произвольное множество, а $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$. Множество всех раскрасок обозначается через $C^{I}$.

Lemma 2
Пусть $G$ группа, действующая на множестве $I$. Тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G$, $i \in I$, а $f \in C^{I}$ задает действие $G$ на $C^{I}$.

Lemma 3
Для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:

1)$gf = f$

2)$f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$

3)Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$ (в терминологии задачи о бусах, это означает, что все места из одного цикла покрашены в одинаковый цвет)

Corollary 1
Количество раскрасок из $C^{I}$, которые сохраняет данный элемент $g \in G$ равно $|C^{c(\varphi(g))}|$, где $c(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$ (считая циклы длины 1).

### Аналитическое решение задачи

Заномеруем грани тетраэдра числами от 1 до 4. Тогда раскраска граней тетраэдра - сопоставление каждому из этих чисел некоторого цвета из $\{x1, x2, x3, x4\}$, где $x \in \{a, b, c, d\}$, где a, b, c, d - некоторые разные цвета. Группа G вращений тетраэдра с операцией композиции состоит из вращений вокруг четырех осей, соединяющих вершины с центрами противоположных граней (8 вращений), вращений на угол вокруг каждой из трех осей, соединяющих середины противоположных ребер (3 вращения) и тождественного преобразования.

Итак, группа G действует на множестве раскрасок X. Раскраски граней тетраэдра, совмещаемые друг с другом каким-нибудь преобразованием группы G , лежат в одной орбите под действием группы G. Определить количество различных раскрасок, означает найти число орбит действия группы G на множестве X. Для этого воспользуемся леммой Бернсайда. Выясним, сколько раскрасок из X оставляет на месте каждый элемент группы G.  
Нейтральный элемент e оставляет на своем месте все раскраски, и поскольку их всего $4^4$ , то и $|X^e| = 4^4$. Вращение $\tau_1^1 = (1)(243)$ сохраняет раскраску, если грани с номерами 2, 3, 4 окрашены в один и тот же цвет, цвет же первой грани при этом может быть любым. Значит, первая грань может быть окрашена четырьмя способами, и в то же время оставшиеся грани вместе красятся в один из четырёх цветов, поэтому $|X^{\tau_1^1}| = 4^2$. Поскольку остальные вращения $\tau_k^i$ имеют подобную циклическую структуру, то $|X^{\tau_1^2}| = |X^{\tau_2^1}| = |X^{\tau_2^2}| = |X^{\tau_3^1}| = |X^{\tau_3^2}| = |X^{\tau_4^1}| = |X^{\tau_4^2}| = 4^2$. Подстановка $\sigma_1 = (13)(24)$ сохранит раскраску граней при условии, что грани 1 и 3 выкрашены в один цвет (четыре возможности) и в то же время грани 2 и 4 выкрашены в один цвет (также четыре возможности). Значит, $|X^{\sigma_1}| = 4^2$. Аналогичным образом имеем
$|X^{\sigma_2}| = |X^{\sigma_3}| = 4^2$. Учитывая, что порядок группы G равен 12, по лемме Бернсайда находим  
$$N = \frac{1}{|G|} \sum\limits_{g \in G}{|X^g|} = \frac{4^4 + 11 \cdot 4^2}{12} = 36$$

### Реализация функции, графически отображающей раскраску по заданному номеру

Функции случайного распределения четырёх цветов между гранями и создания тетраэдра:

```python
import random

def random_colors():
    colors = []
    for i in range(4):
        colors.append(random.randint(1,4))
    return colors

colors = random_colors()
print(colors)
```

```python
import math
from sage.plot.plot3d.shapes import Text

def create_tetrahedron(colors):  
    colorDict = {1 : 'blue', 2 : 'green', 3 : 'yellow', 4 : 'red'}
    T1 = polygon3d([(0,0,0), (math.sqrt(3)/2, 0.5, 0), (0,1,0)], color = colorDict[colors[0]])
    T2 = polygon3d([(math.sqrt(3)/2, 0.5, 0), (0,1,0), (math.sqrt(3)/6, 0.5, math.sqrt(2)/math.sqrt(3))], color = colorDict[colors[1]])
    T3 = polygon3d([(0,0,0), (0,1,0), (math.sqrt(3)/6, 0.5, math.sqrt(2)/math.sqrt(3))], color = colorDict[colors[2]])
    T4 = polygon3d([(0,0,0), (math.sqrt(3)/2, 0.5, 0), (math.sqrt(3)/6, 0.5, math.sqrt(2)/math.sqrt(3))], color = colorDict[colors[3]])
    faces_numbers = Text("1", color = 'black').translate(0.3,0.5,-0.1) + Text("2", color = 'black').translate(0.5, 0.2, 0.3) + Text("3", color = 'black').translate(0.5, 0.7, 0.3) + Text("4", color = 'black').translate(0,0.5,0.5)
    return T1 + T2 + T3 + T4 + faces_numbers
T = create_tetrahedron(colors)
T.show()
```

Вращение относительно середины грани 1 и противоположной вершины (120 градусов)

```python
new_colors = [colors[0], colors[3], colors[1], colors[2]]
new_T1 = create_tetrahedron(new_colors)
T.show()
new_T1.show()
```

## Решение задачи в общем виде

Пусть n - количество цветов, тогда 
1. $|X^e| = n^4$;  
2. $|X^{\tau_i^k}| = n^2, где i = 1,..,4; k = 1, 2$;  
3. $|X^{\sigma_i}| = n^2, где i = 1, 2, 3$;  
По лемме Бернсайда:  
$N = \frac{1}{|G|} \sum\limits_{g \in G}{|X^g|} = \frac{n^4 + 8 \cdot n^2 + 3 \cdot n^2}{12} = \frac{n^4 + 11 \cdot n^2}{12}$
