## Теоретические положения

---

Пусть $G$ - группа с нейтральным элементом $e$, $X$ - множество.

$G$ действует на $X$, если задана операция: $G\times X \rightarrow X$, обозначаемая $gx$, где $x\in X$, $g \in G$
Свойства:
* $ex = x$, $\forall x \in X$
* $g(hx) = (gh)x$, $\forall x \in X$, $\forall g,\,h \in G$

_Орбитой_ элемента $x \in X$ под действием $G$ называется множество $Orb(x) = \{ gx\,|\,g \in G\}$. Количество элементов в орбите называется _длиной_ орбиты.

_Стабилизатором_ элемента $x \in X$ называется множество $St(x) = \{ gx=x\,|\,g \in G \}$. 

Длину орбиты любого элемента можно выразить следующим образом: $|Orb(x)| = \frac{|G|}{|St(x)}$.

_Неподвижными точками_ элемента $g \in G$ назвается множество $X^g = \{ gx = x\,|\,x \in X \}$.

___Теорема Бёрнсайда___

Количество орбит действия $G$ на $X$ равно $N = \frac{1}{|G|}\sum_{g \in G} X^g$

Пусть $I$ - произвольное множество, $C$ - множество цветов. _Раскраской_ множества $I$ называется функция из $I$ в $C$. $C^I$ - множество всех раскрасок. $|C^I|=|C|^{|I|}$

Пусть $G$ - группа, действующая на $I$. Тогда формула $gf(i)=f(g^{-1}i)$ задаёт действие $G$ на $C^I$, $g \in G$, $i \in I$, $f \in C^I$.

Для элемента $g \in G$, $f \in C^I$ эквивалентны следующие утверждения:
1. $gf = f$
2. $f(i) = f(g^n i)$, $\forall i \in I$, $\forall n \in Z$
3. Если элементы $i,\,j \in I$ лежат в одном цикле в циклической записи перестановок $\phi(g)$, то $f(i)=f(j)$

Количество раскрасок из $C^I$, которое сохраняет элемент $g \in G$ равно $|C|^{c(\phi(g))}$, где $c$ - количество независимых циклов в перестановке, включая циклы единичной длины.

### Графическое отображение

Функции, отображающие раскраску и выпоняющие повороты и осевые симметрии.

```python
from random import randint

colors = ["red", "green", "blue", "orange", "purple"]

n = 9
k = len(colors)

def generate_number():
    number = 0
    ratio = 1
    for i in range(n**2):
        number += ratio * randint(0, k - 1)
        ratio *= k
    return number

def get_matrix_from_number(number):
    matrix = []
    for i in range(n):
        line = []
        for j in range(n):
            line.append((i * n + j + 1, number % k))
            number //= k
        matrix.append(line)
    return matrix

def get_random_matrix():
    return [[(i * n + j + 1, randint(0, k - 1)) for j in range(n)] for i in range(n)]

if __name__ == "__main__":
    print("Number:", generate_number())
```

```python
from sage.plot.graphics import Graphics
from sage.plot.polygon import polygon
from sage.plot.text import Text

def print_matrix(matrix):
    g = Graphics()
    for i in range(n):
        for j in range(n):
            color = matrix[i][j][1]
            number = matrix[i][j][0]
            g += polygon([[j, n - i], [j + 1, n - i], [j + 1, n - i - 1], [j, n - i - 1]],
                            color=colors[color],
                            edgecolor="black", thickness=3)
            g += text(number, (j + 0.5, n - i - 0.5), color="black", fontweight="bold")
    g.show(figsize=[7,7], axes=False)

if __name__ == "__main__":
    number = int(input("Number: "))
    print_matrix(get_matrix_from_number(number))
```

```python
def matrix_rotate90(matrix):
    for line in range(n//2):
        for i in range(line, n - line - 1):
            (matrix[i][n - line - 1], matrix[n - line - 1][n - i -1], matrix[n - i - 1][line],
            matrix[line][i]) = (matrix[line][i], matrix[i][n - line - 1],
            matrix[n - line - 1][n - i -1], matrix[n - i - 1][line])
    return matrix
  
def matrix_rotate180(matrix):
    for i in range((n**2)//2):
        col = i % n
        line = i // n
        matrix[line][col], matrix[n - line - 1][n - col - 1] = matrix[n - line - 1][n - col - 1], matrix[line][col]
    return matrix

def matrix_rotate270(matrix):
    for line in range(n//2):
        for i in range(line, n - line - 1):
            (matrix[line][i], matrix[i][n - line - 1], matrix[n - line - 1][n - i -1],
            matrix[n - i - 1][line]) = (matrix[i][n - line - 1], matrix[n - line - 1][n - i -1],
            matrix[n - i - 1][line], matrix[line][i])
    return matrix

if __name__ == "__main__":
    number = int(input("Number: "))
    print("Исходная матрица:")
    print_matrix(get_matrix_from_number(number))
    print("Поворот на 90:")
    print_matrix(matrix_rotate90(get_matrix_from_number(number)))
    print("Поворот на 180:")
    print_matrix(matrix_rotate180(get_matrix_from_number(number)))
    print("Поворот на 270:")
    print_matrix(matrix_rotate270(get_matrix_from_number(number)))
```

```python
def sym_vertical(matrix):
    for i in range(n):
        for j in range(n//2):
            matrix[i][j], matrix[i][n - j - 1] = matrix[i][n - j - 1], matrix[i][j]
    return matrix

def sym_horizontal(matrix):
    for i in range(n//2):
        for j in range(n):
            matrix[i][j], matrix[n - i - 1][j] = matrix[n - i - 1][j], matrix[i][j]
    return matrix

def sym_main_diagonal(matrix):
    for i in range(n):
        for j in range(i, n):
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
    return matrix

def sym_antidiagonal(matrix):
    for i in range(n):
        for j in range(n - i):
            matrix[i][j], matrix[n - j - 1][n - i - 1] = matrix[n - j - 1][n - i - 1], matrix[i][j]
    return matrix

if __name__ == "__main__":
    number = int(input("Number: "))
    print("Исходная матрица:")
    print_matrix(get_matrix_from_number(number))
    print("Симметрия по вертикали:")
    print_matrix(sym_vertical(get_matrix_from_number(number)))
    print("Симметрия по горизонтали:")
    print_matrix(sym_horizontal(get_matrix_from_number(number)))
    print("Симметрия по главной диагонали:")
    print_matrix(sym_main_diagonal(get_matrix_from_number(number)))
    print("Симметрия по дополнительной диагонали:")
    print_matrix(sym_antidiagonal(get_matrix_from_number(number)))
```

### Решение в общем виде

---

Пусть $n$ - количество частей, на которые разбита сторона квадрата, $k$ - количество цветов.

Суммарное количество всех поворотов и симметрий равно $8$, т.е. $|G| = 8$.

__Если $n$ - нечётное__:

Центральная клетка всегда будет оставаться на месте. При повороте перемещаются $(\frac{n-1}{2} + 1)\frac{n-1}{2}$ клеток. Из этого получаем, что количество циклов в $\tau$ и $\tau^3$ равно $\frac{n^2-1}{4} + 1$.

При повороте на $180^{\circ}$ длина циклов уменьшается в два раза, исключая центральный, значит, их количество увеличивается в два раза. Из этого получаем, что количество циклов в $\tau^2$ равно $\frac{n^2-1}{2} + 1$.

При всех симметриях $n$ элементов останутся на местах. Остальные элементы в паре с другим образуют цикл. Из этого получаем, что количество циклов во всех $\sigma$ равно $\frac{n^2-n}{2}+n$.

Получаем следующее:
$$
\begin{array}{l}
|X^{id}| = k^{n^2}\\
|X^{\tau}| = k^{\frac{n^2-1}{4} + 1}\\
|X^{\tau^2}| = k^{\frac{n^2-1}{2} + 1}\\
|X^{\tau^3}| = k^{\frac{n^2-1}{4} + 1}\\
|X^{\sigma_1}| = k^{\frac{n^2-n}{2}+n}\\
|X^{\sigma_2}| = k^{\frac{n^2-n}{2}+n}\\
|X^{\sigma_3}| = k^{\frac{n^2-n}{2}+n}\\
|X^{\sigma_4}| = k^{\frac{n^2-n}{2}+n}\\
\end{array}
$$

Итоговая формула для количества вариантов с точностью до поворотов и симметрий будет иметь следующий вид:
$$
N = \frac{1}{|G|}\sum_{g \in G} |X^g| = \frac{1}{8}(k^{n^2}+k^{\frac{n^2-1}{2} + 1}+2k^{\frac{n^2-1}{4} + 1}+4k^{\frac{n^2-n}{2}+n})
$$

__Если $n$ - чётное__:

При повороте каждая четверть будеть переходить в другую, значит, количество циклов для $\tau$ и $\tau^3$ будет равно $\frac{n^2}{4}$.

При повороте на $180^{\circ}$ длина циклов уменьшается в два раза, значит, их количество увеличивается в два раза. Из этого получаем, что количество циклов в $\tau^2$ равно $\frac{n^2}{2}$.

При симметрии по горизонтали или вертикали одна половина перейдёт в другую. Значит, количество циклов для $\sigma_1$ и $\sigma_2$ будет равно $\frac{n^2}{2}$. При симметрии по диагонали $n$ элементов останется на месте, а остальные попарно будут переходить в друг друга. Значит, количество циклов для $\sigma_3$ и $\sigma_4$ будет равно $\frac{n^2-n}{2}+n$.

Получаем следующее:
$$
\begin{array}{l}
|X^{id}| = k^{n^2}\\
|X^{\tau}| = k^{\frac{n^2}{4}}\\
|X^{\tau^2}| = k^{\frac{n^2}{2}}\\
|X^{\tau^3}| = k^{\frac{n^2}{4}}\\
|X^{\sigma_1}| = k^{\frac{n^2}{2}}\\
|X^{\sigma_2}| = k^{\frac{n^2}{2}}\\
|X^{\sigma_3}| = k^{\frac{n^2-n}{2}+n}\\
|X^{\sigma_4}| = k^{\frac{n^2-n}{2}+n}\\
\end{array}
$$

Итоговая формула для количества вариантов с точностью до поворотов и симметрий будет иметь следующий вид:
$$
N = \frac{1}{|G|}\sum_{g \in G} |X^g| = \frac{1}{8}(k^{n^2}+2k^{\frac{n^2}{4}}+3k^{\frac{n^2}{2}}+2k^{\frac{n^2-n}{2}+n})
$$

Общая формула имеет следующий вид:
$$
N = \begin{cases}
\frac{1}{8}(k^{n^2}+k^{\frac{n^2-1}{2} + 1}+2k^{\frac{n^2-1}{4} + 1}+4k^{\frac{n^2-n}{2}+n}),\,\,if\,\,n \bmod 2 = 1 \\
\frac{1}{8}(k^{n^2}+2k^{\frac{n^2}{4}}+3k^{\frac{n^2}{2}}+2k^{\frac{n^2-n}{2}+n}),\,\,if\,\,n \bmod 2 = 0
\end{cases}
$$
