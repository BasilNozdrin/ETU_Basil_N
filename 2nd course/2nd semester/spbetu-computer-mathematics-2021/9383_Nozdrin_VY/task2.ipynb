**Постановка задачи:** Круг разбит на $154$ сектора, каждый из которых покращен в один из $6$ цветов. Сколькими способами можно составить такую мозаику (с точностью до поворотов круга)?  

### Определение  

Пусть $G$ - группа с нейтральным элементом $e$, и $X$ - некоторое множество.  
Говорят, что $G$ действует на $X$, если задана операция $G\times X\to G$  
Со следующими свойствами:  
1. $\forall x\in X:\quad ex=x$  
2. $\forall x\in X,\forall g,h\in G:\quad g(hx)=(gh)x$

### Определение  

Орбитой $x\in X$ по действию группы $G$ на множество $X$ называется множество $Gx = \{gx|g\in G\}$.  

Длина орбиты - количество элементов в данной орбите.  

### Определение  

Неподвижными точками элемента $g\in G$ называют те $x\in X$ для которых $gx = x$.  

$X^g \quad-$ множество неподвижных точек по элементу $g$  

### Лемма Бёрнсайда

Количество орбит действия группы $G$ на множестве $X$ равно $\frac{1}{|G|}\sum_{g\in G}|X^g|$


### Определение  

Пусть $I$ - произвольное множество, и $C$ - множество цветов.  
Раскраской множества $I$ называется функция из $I$ в $C$.  
Множество всех раскрасок обозначается $C^{I}$.  
Количество раскрасок равно ${|C|}^{|I|}$

### Определение  

Если $G$ - группа, действующая на множестве $I$, тогда для любых $g\in G, i\in I, f \in C^I$ 
выражение $gf(i)=f(g^{-1}i)$ задает действие $g$ на $C^I$. 

### Теорема   

Для любого $g\in G$ и любой функции $f \in C^I$ следующие условия эквивалентны:  
1. $gf=f$
2. $\forall i\in I,\forall n\in\mathbb Z\quad f(i)=f(g^ni)$
3. $\forall i,j \in I$ если $i, j$ лежат в одном цикле в циклической записи перестановки $\varphi(x)$, то $f(i)=f(j)$. (В терминологии задачи о бусах это означает, что все места из одного цикла покрашены в одинаковый цвет).  

### Определение  

Количество раскрасок $C^I$, которые сохранят данный элемент $g\in G$ равно ${|C|}^{c\varphi(g)}$.  
где $c\varphi$ - количество независимых циклов в перестановке $\varphi$.

## Выполнение работы  

**Постановка задачи:** Круг разбит на $154$ сектора, каждый из которых покращен в один из $6$ цветов. Сколькими способами можно составить такую мозаику (с точностью до поворотов круга)?  

### Решение  

Рассмотрим $\{a,b,c,d,e,f\}\quad-$ множество цветов, $\{x_1,x_2,x_3\dots,x_{154}\}\quad-$ множество секторов круга, где $\forall i\  x_i \in \{a,b,c,d,e,f\}$. Всего таких мозаик будет $6^{154}$ не учитывая повороты круга.

На множестве всех раскрасок (мозаик) действует группа поворотов с операцией композиции $(G,\circ)$.  
В этой группе нейтральный элемент $e=\mathbb{1}_X$ и все остальные элементы являются степенями циклической перестановки $\tau=(1,2,3,\dots,154)$. $\tau^{154}=e$.

Раскраски (мозаики), совмещенные друг с другом преобразованиями из $G$ задают одинаковые расскраски, другими словами раскраски совпадают, если они лежат в одной орбите по элементам группы $G$. Вычислим количество орбит действия $G$ на множестве раскрасок. 

Выпишем некоторые перестановки из $S_{154}$ в которые переходят элементы группы $G$:

$\tau^0 =
\begin{pmatrix}
  1 & 2 & 3 & \dots & 153 & 154\\
  1 & 2 & 3 & \dots & 153 & 154
\end{pmatrix}=(1)(2)(3)\dots(154)=e\\
\tau^1 =
\begin{pmatrix}
  1 & 2 & 3 & \dots & 153 & 154\\
  2 & 3 & 4 & \dots & 154 & 1
 \end{pmatrix}=(1,2,3,\dots,154)\\
\tau^2 =
\begin{pmatrix}
  1 & 2 & 3 & \dots & 153 & 154\\
  3 & 4 & 5 & \dots & 1   & 2
\end{pmatrix}=(1,3,5,\dots,153)(2,4,\dots,154)$ 

Перестановка $\tau$ является циклом длины 154. $\forall n \tau^n$ можно вычислить количество циклов в перестановке. Число $154$ раскладывается на простые множители как $154=2*7*11$. Каждая перестановка $\tau^n$ является произведением $d$ независимых циклов, где $d:=$НОД$(154,n)$. Следовательно для всех $n$ не кратных 2, 7, или 11, $d$ будет равняться 1.  
Теперь мы можем найти количество раскрасок, которые сохраняет данный элемент $g \in G$:

$|X|e=6^{154}\\
|X|{\tau^k}=6,         \forall k\in\{x\in\mathbb{N}| 1\leq x \leq154, 2\nmid x, 7\nmid x, 11\nmid x, \}\\
|X|{\tau^{2*l}}=6^2,    \forall l\in\mathbb{N}\\
|X|{\tau^{7*l}}=6^7,    \forall l\in\mathbb{N}\\
|X|{\tau^{11*l}}=6^{11},\forall l\in\mathbb{N}\\
|G|=154$

По лемме Бернсайда количество различных мозаик равно $\frac{1}{154}(6^{154}+6*60+6^{2}*77+6^{7}*22+6^{11}*14)$

### Графическое представление

```python
import matplotlib.pyplot as plt
import random
```

```python
size = 154
colours = ['red','orange','yellow','green','cyan','blue']
sizes = [1 for i in range(size)]

def getMosaic(n):
    pie_colours = []
    for i in range(size):
        pie_colours.append(colours[n%6])
        n //= 6
    return pie_colours
```
