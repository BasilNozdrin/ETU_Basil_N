## Основные теоретические положения
_**Бинарный метод возведения числа в степень**_

Бинарное (двоичное) возведение в степень — это приём, позволяющий возводить любое число в $n$-ую степень за $O(\log n)$ умножений (вместо $n$ умножений при обычном подходе).

Для любого числа $a$ и чётного числа $n$ выполнимо тождество (следующее из ассоциативности операции умножения):

 $a^n = (a^{n/2})^2 = a^{n/2} \cdot a^{n/2}$

Действительно, для чётного $n$ мы показали, как, потратив всего одну операцию умножения, можно свести задачу к вдвое меньшей степени.
Если степень $n$ нечётна, перейдём к степени $n-1$, которая будет уже чётной:

 $a^n = a^{n-1} \cdot a$

Итак, от степени $n$ мы переходим, если она чётна, к $n/2$, а иначе — к $n-1$. Понятно, что всего будет не более $2 \log n$ переходов, прежде чем мы придём к $n=0$ (базе рекуррентной формулы). Таким образом, мы получили алгоритм, работающий за $O(\log n)$ умножений. Количество операций умножения в бинарном методе равно $\lambda(n) + \nu(n) - 1$, где $\lambda(n) = \lfloor\log_2 n\rfloor, \nu(n)$ - количество единиц в бинарном представлении числа.

_**Метод множителей для $x^n$**_

Метод множителей для $x^n$ - это метод возведения числа в степень $n \in N$. В случае если $n$ - составное число, то есть $i \cdot j = n$, где $i$ - наименьший простой множитель $n$, то $x^n = x^{i \cdot j} = (x^i)^j$.

В случае если $n$ - простое число, то $x^n = x^{(k-1) + 1} = x^{k-1} \cdot x$.

_**Адитивная цепочка**_

Аддитивной цепочкой для числа $N$ называется
последовательность натуральных чисел $1, 2, a_2, a_3, \ldots, a_l = N$, где
каждый следующий член получается путем сложения каких-либо двух
предыдущих: 

$a_i = a_j + a_k,  0 ≤ k ≤ j < i,  i = 1, 2, …, l,$ при этом число $l$ называется длиной аддитивной цепочки.

_**Алгоритм Яо**_

Алгоритм Яо - алгоритм вычисления аддитивной цепочки для заданного $k \in N$. Пусть даны некоторые два числа $ n,k \in N$, тогда $n$ можно представить как: $2^{jk} \cdot c_j + \ldots + 2^{2k} \cdot c_2 + 2^{k} \cdot c_1 + c_0$, где $c_j \not = 0$. 
Пусть $d(z)$ - это сумма чисел $2^{ik}$ для всех $i$ таких, что $c_i = z$.

Цепочка Яо начинается с $1, 2, 4, 8,  \ldots , 2^{\lfloor\lg n\rfloor}$, складываем различные степени двойки вместе, чтобы получить $d(z)$, для любого $z \in \{1, 2, \ldots, 2 ^ {k-1} \}$, где $d(z) \neq 0$. Затем получив $z \cdot d(z)$ для каждого $z$, получим $n = d(1) + 2 \cdot d(2) + 3 \cdot d(3) + \ldots + (2^k - 1) \cdot d(2^k - 1)$.

_**Алгоритм дробления вектора индексов**_

Звездная цепочка - это аддитивная цепочка, в которой каждый элемент подчиняется определенному правилу: $a_i=a_{i-1}+a_j$, где $1 \leq j \leq i$

Для каждой звездной цепочки существует вектор индексов - последовательность натуральных чисел, где каждый ее элемент является индексом $j$ в $a_j$ из формулы $a_i=a_{i-1}+a_j$.

Алгоритм дробления вектора индексов заключается в поиске звездной цепочки минимальной длины для заданного числа $n \in N$.

Вначале рассмотрим аддитивные цепочки длины $m$, где $\underline{l}(n) \leq m \leq \overline{l}(n)$, $\underline{l}(n) = \lfloor lb n\rfloor$ и $\overline{l}(n) = \lambda(n) + v(n) - 1$.

Выберем $q \in N$ такое, что $1 \leq q \leq m - 1$, пусть $q = \frac{m}{2}$. Переберем векторы индексов длины $q$, они будут указывать на первые элементы в цепочке, их будем фиксировать. Чтобы вектор индексов был размером $m$, дополним его двумя возможными способами:

* Чтобы получить в цепочке последний минимальный элемент, добавляем к нашему вектору вектор индексов длины $m - q$, содержащий только единицы в качестве элементов.
* Чтобы получить в цепочке последний максимальный элемент, добавляем к нашему вектору вектор индексов длины $m - q$, содержащий элементы, которые будут равны своим индексам при нумерации, начинающейся с 1.
    
Таким образом мы получаем $a_{max}$ и $a_{min}$.

При равенстве $n$ и $a_{max}$ или $a_{min}$ алгоритм заканчивается.

При $n \in [a_{min}, a_{max}]$ перебираем вектора индексов длины $m - q$, чтобы при добавлении к нашему вектору по получившемуся можно было бы построить аддитивную цепочку с последним элементом $n$.

При $n \not \in [a_{min}, a_{max}]$ возьмем следующий вектор индексов длинной $q$ вместо предыдущего фиксированного.


_**Следствие 1 теоремы Брауэра**_

Теорема Брауэра. Для числа $k < lb (lb n)$ справедливо: $l(n) < (1+k^{-1}) \cdot \lceil lb n \rceil + 2^{k-1} - k + 2$.

**1. Бинарный метод и метод множителей**

Ниже представлена реализация бинарного и метода множителей:

```python
def binary_method(a, n):
    res = a
    op = 0
    chain = [1]
    bin_n = bin(n)[3::]
    while len(bin_n):
        if (bin_n[0] == '1'):
            res *= res * a
            chain.append(chain[len(chain) -1] * 2 + 1)
            op += 2
        else:
            res *= res
            chain.append(chain[len(chain) -1] * 2)
            op += 1
        bin_n = bin_n[1:]
    return [res, op, chain]
```

```python
def min_simple_div(k):
    i = 2
    while i*i <= k:
        if (k % i == 0):
            return i
        i += 1
    return k

def multiplier_method(a, n):
    msd = min_simple_div(n)
    if (msd == n):
        res = binary_method(a, n - 1)
        res[0] *= a
        res[2].append(chain[len(chain) -1] + 1)
        res[1] += 1
        return res
    else:
        chain =  [[] for i in range(2)]
        res = binary_method(a, msd)
        chain[0] = res[2]
        op = res[1]
        res = binary_method(res[0], n // msd)
        chain[1] = res[2]
        res[1] += op
        res[2] = chain
        return res
```

**2. Алгоритм Яо**

Ниже представлена реализация алгоритма Яо:

```python
def algorithm_Yao(n, k):
    new_count,chain = [], []
    d_z = {}
    c, m, base = -1, n, 2**k
    while m:
        new_count.append(m % base);
        m >>= k
        c += 1
    for power in range(c * k + 1):
        chain.append(2**power)
    for i in range(len(new_count)):
        if new_count[i] in d_z:
            d_z[new_count[i]] += base**i
        else:
            d_z[new_count[i]] = base**i
    sorted_d_z = dict(sorted(d_z.items(), key=lambda x: x[0]))
    sum_d_z = 0
    for key, value in sorted_d_z.items():
        t, d = bin(key), 0
        t = t[2:]
        while len(t):
            if (t[0] == '1'):
                d += value
                if d not in chain:
                    chain.append(d)                    
            if (len(t) != 1):
                d += d
                if d not in chain:
                    chain.append(d)
            t =t[1:]
        sum_d_z += d
        if sum_d_z not in chain and sum_d_z != 0:
            chain.append(sum_d_z)
    return chain
```

Применим алгоритм Яо для построения аддитивных цепочек для заданных $n$: $19, 51, 99, 123, 757$

**3. Алгоритм дробления вектора индексов**

Ниже представлена реализация алгоритма дробления вектора индексов:

```python
import math, time

def to_bin(n):
    b = bin(n)
    return b[2::]

def get_star_chain(vec_ind):
    s_c = [1]
    for i in range(1, len(vec_ind) + 1):
        s_c.append(s_c[i - 1] + s_c[vec_ind[i - 1] - 1])
    return s_c

def get_new_vec(vec, q):
    length = len(vec)
    flag = false
    for i in range(length - 1, -1, -1):
        if (vec[i] != 1):
            vec[i] -= 1
            flag = true
            break;
        else:
            vec[i] = q + i + 1
    return vec, flag

def get_vec(q):
    return [i + 1 for i in range(q)]
    
def get_min_vec(m):
    return [1 for i in range(m)]
    
def get_max_vec(m, q):
    return [q + i + 1 for i in range(m)]

def algorithm_dr_vec_ind(n):
    start_time = time.time()
    for m in range(int(math.log(n, 2)), int(math.log(n,2))+ int(to_bin(n).count('1'))):
        q = m // 2
        if q == 0:
            q = 1
        vec, flag = get_vec(q), true
        while (flag):
            min_c = get_star_chain(vec + get_min_vec(m - q))
            max_c = get_star_chain(vec + get_max_vec(m - q, q)) 
            min_c_el, max_c_el = min_c[len(min_c) - 1], max_c[len(max_c) - 1]
            if (min_c_el == n):
                return [min_c, time.time() - start_time]
            if (max_c_el == n):
                return [max_c, time.time() - start_time]
            if n in range(min_c_el, max_c_el):
                cur_vec = get_max_vec(m - q, q) 
                cur_c = [1]
                c_flag = true
                while (cur_c[len(cur_c) - 1] != n):
                    cur_vec, c_flag = get_new_vec(cur_vec, q)
                    if not c_flag:
                        break
                    cur_c = get_star_chain(vec + cur_vec) 
                if (c_flag):
                    return [cur_c, time.time() - start_time]
            vec, flag = get_new_vec(vec, 0)
    return [[1], time.time() - start_time]
```

Сравним алгоритм Яо и алгоритм дробления вектора индексов для заданных $n$: $555, 765, 808$

```python
from matplotlib import pyplot as plt
headers = ['n', 'time', 'star chain', 'chain of dr vec ind', 'Yao chain']
list_of_n = [555, 765, 808]
x, y = len(list_of_n), 5
data = [[0 for j in range(y)] for i in range(x)]
for i in range(len(list_of_n)):
    al = algorithm_dr_vec_ind(list_of_n[i])
    data[i][0] = list_of_n[i]
    data[i][1] = al[1]
    data[i][2] = al[0]
    data[i][3] = len(data[i][2])
    data[i][4] = len(algorithm_Yao(list_of_n[i], 2))
    
table = plt.table(cellText=data, rowLoc='right', colLabels=headers, loc='center')

table.auto_set_font_size(False)
table.auto_set_column_width(col=list(range(4)))
table.set_fontsize(14)
table.scale(2, 3) 


ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)
print('Сравнение алгоритма Яо и алгоритма дробления вектора индексов:')
plt.draw()
```

**Вывод**

Для алгоритма Яо k = 2, так как в большей степени при этом значении он выдает наименьшую возможную аддитивную цепочку по длине. Исходя из таблицы выше можно утверждать, что алгоритм дробления вектора индексов выдает минимальную аддитивную цепочку, чем алогритм Яо, но затрачивает больше ресурсов на вычисления.

**4. Следствие 1 теоремы Брауэра**

Ниже представлена реализация проверки следствия 1 теоремы Брауэра:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import leastsq

def model(t, k):
    return np.log(t) * k[0]

def residuals(k, y, t):
    return  y - model(t, k)

n = 200
x, y = [i for i in range(1, n + 1)], [len(algorithm_dr_vec_ind(i)[0]) for i in range(1, n + 1)]
figure, axes = plt.subplots(figsize = (15,15))
p, flag = leastsq(residuals, 1, args=(y, x))
plt.plot(x, y, 'o',  x, model(x, p))
plt.show()
```

**Вывод**

Исходя из графика выше, можно утрверждать, что построение похоже на логарифмическую кривую, что подтверждает теорему Брауэра. При увеличении значения $n$ длина аддитивной цепочки будет возрастать все меньше со временем, поэтому на бесконечности она будет эквивалентна $\lambda(n)$
