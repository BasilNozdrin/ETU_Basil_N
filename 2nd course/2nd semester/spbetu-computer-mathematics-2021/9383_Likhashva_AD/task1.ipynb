## Основные теоритические положения 

#### Бинарный метод возведения числа в степень

Бинарный метод возведения в степень - это алгоритм, который позволяет возводить любое число $x$ в степень $n \in N$. В данном методе степень представляется в двоичной системе счисления. Далее происходит просмотр бит слева направо, не учитывая первый бит, так как он всегда равен единице: если бит равен единице, то текущее число возводится в квадрат и умножается на исходное чило, если бит равен нулю, то текущее число возводится в квадрат.
Количество операций умножения в бинарном методе равно: $\lambda(n) + \nu(n) - 1$, где $\lambda(n) = \lfloor log_2 n \rfloor, \ \nu(n)$ - количество единиц в двоичной записи числа.

#### Метод множителей возведения числа в в степень 

Метод множителей возведения числа в в степень - это алгоритм, который позволяет возводить любое число $x$ в степень $n \in N$. Алгоритм следующий: если число простое, то это число возводится в степень $k-1$, далее оно домножается на исходное. Если же число составное, тогда степень числа представляется как произведение $k = i\cdot j$, где $i$ - наименьший простой множитель числа $k$. После число возводится в степень $i$, а затем полученное число возводится в степень $j$.

#### Аддитивная цепочка

Аддитивная цепочка для $n \in N$ - это последовательность $1=a_0, a_1, a_2, ... , a_m = n$, где $a_i = a_j + a_k, k \leq j < i, \forall i = \overline{1...m}$. 
Длина минимальной аддитивной цепочки для числа $n$ равняется $l(n) = m$
Рекуррентная формула длины аддитивной цепочки: $l(n) = min\{l(n - 1) + 1, \ l_n\} + \sigma_n $

#### Теорема Брауэра

Для числа $k < log_2 (log_2 n)$ справедливо, что: $l(n) < (1+k^{-1}) \cdot \lceil log_2 n \rceil + 2^{k-1} - k + 2$.
Если $k=\lambda(\lambda(n))-2\lambda(\lambda(\lambda(n)))$, то следует что:

Следствие 1: $\lim \limits_{n \to \infty} \frac{l(n)}{\lambda(n)} = 1$

Следствие 2: Одна из наилучших верхних оценок длины аддитивной цепочки имеет вид: $\lambda(n)(1 + \frac{1}{\lambda(\lambda(n))} + \frac{o(\lambda(\lambda(\lambda(n))))}{\lambda(\lambda(n)) ^ 2})$

#### Алгоритм Брауэра

Алгоритм Брауера вычисляет n-ую стпенень за $\lambda(n)+\frac{(1+o(1))\lambda(n)}{\lambda(\lambda(n))}$ операций.

Для некоторых n, k Брауерские цепочки задаются в виде рекурентной формулы: $B_k(n) =\begin{cases}1, 2, 3, ..., 2^k-1\text{, если } n < 2^k \\ B_k(q), 2q, 4q, 8q, ..., 2^kq, n,\text{ если } n \geqslant 2^k\ \text{и } q = \lfloor\frac{n}{2^k}\rfloor \end{cases}$

#### Звёздная цепочка

Аддитивная цепочка называется звездной, если эта аддитивная цепочка включает в себя только звездные шаги($a_i = a_{i-1} + a_k, k = \overline {0...(i-1)}$).

#### Алгоритм дробления вектора индексов

Для данного $n \in N$ надо найти $\min l^*(n) = m: a_m = n$
Алгоритм:
1) Во внешнем цикле рассматриваются аддитивные цепочки длины $m$: $\underline{l}(n) \leq m \leq \overline{l}(n)$, где $1 \leq q \leq m - 1$, пусть $q = \frac{m}{2}$, $q \in N$

2) Во внутреннем цикле перебираются все фиксированные части ВИ $\{r_i\}_{i=1}^{q}$ ($q!$ шагов) на каждом шаге строим $\{r_i\} \to \{a_i\}$, вычисляем $a_{min}$ и $a_{max}$
 * если на каком-то шаге $a_m = n \ \to $ решено
 * если $n \notin [a_{min}, a_{max}] \to $ переходим к следующему набору $\{r_i\}_{i=1}^{q}$
 * если $n \in [a_{min}, a_{max}] \to $ организуем внутренний цикл перебора меняющейся части $\{\rho_i\}_{q+1}^{m}$
 * если $\rho_i = 1 \ \forall j \ \to $ переходим с слудующему $\{r_i\}_{i=1}^{q}$

3) Если $r_i = 1 \ \forall i$ и $\rho_i = 1 \ \forall j \ \to m = m + 1$

#### Гипотеза Шольца - Брауэра

$l^*(2^n-1) \leq l^*(n)+n-1$

### 1. Бинарный метод и метод множителей

#### Бинарный метод

```python
def DecToBin(x):
    ans = bin(x)
    return ans[3:]

def BinaryMethod(x, n):
    step = 0
    res = x
    temp = DecToBin(n)
    degSeq = [1]
    for j in temp:
        if j == '1':
            res *= res
            degSeq.append(degSeq[-1] + degSeq[-1])
            res *= x
            degSeq.append(degSeq[-1] + 1)
            step += 2
        elif j == '0':
            res *= res
            degSeq.append(degSeq[-1] + degSeq[-1])
            step += 1
    return res, step, degSeq

def CountBinaryMethod(x, n): 
    res, step, degSeq = BinaryMethod(x, n)
    return step, degSeq


#stepsBinM, seqBinM = CountBinaryMethod(1, 5)
#print("Шаги бинарного метода:", stepsBinM)
#print("Последовательность степеней для бинарного метода: ", seqBinM)
```

#### Метод множителей

```python
def MinDiv(x):
    i = 3
    while x%i != 0 and i*i <= x:
        i += 2
    if i*i <= x:
        return i
    else:
        return x
    if x%2 == 0:
        return 2

def CountMulMethod(x, n):
    min = MinDiv(n)
    x1, step1, SeqDegX = BinaryMethod(x, min)
    x2, step2, SeqDegY = BinaryMethod(x1, n/min)
    res = step1 + step2
    return res, SeqDegX, SeqDegY

#StepsMul, DegXSeq, DegYSeq = CountMulMethod(1, 5)
#print("Шаги метода множителей:", str(StepsMul))
#print("Последовательность степеней для метода множителей: x: ", DegXSeq, ", y = x^" + str(DegXSeq[-1]) + ": ", DegYSeq)
```

### 2. Алгоритм Брауэра

```python
from matplotlib import pyplot as plt
table_res = []

def AlgorithmBrauer(n, k, add):
    if(n < 2**k):
        for i in range(1, 2**k):
            add.append(i)
        return    
    else:
        q = n//(2**k)
        AlgorithmBrauer(q, k, add)
        for i in range(1, k+1):
            q *= 2 
            add.append(q)    
        add.append(n)
        
minimalChain = {}  #минимальная аддитивная цепочка
n = [10, 22, 73, 141]
k = [2, 3, 4, 5]
for i in n:
    for j in k:
        add = []
        AlgorithmBrauer(i, j, add)
        print("n =", i, ", k =", j)
        print("Аддитивная цепочка: ", add)
        print("Длина аддитивной цепочки: ", len(add))
        print()
        
        if minimalChain.get(i) is None or len(minimalChain[i][1]) >= len(add):
            minimalChain[i] = [j, add]
        
        row = [i, j, len(add)]
        table_res.append(row)
        
headers = ['n', 'k', 'Длина цепочки Брауэра']
cell_text = []
for i in table_res:
    cell_text.append([data for data in i])

table = plt.table(cell_text, rowLoc='right', colLabels=headers, loc='center')
table.set_fontsize(20)
table.scale(4, 2) 

ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)

plt.draw()  
```

### 3. Алгоритм дробления вектора индексов 

```python
import time
import math
from matplotlib import pyplot as plt

def Nu(x):
    return bin(x)[3:].count('1')


def MakeChain(indexV):
    chain = [1]
    for i in indexV:
        chain.append(chain[-1] + chain[i - 1])
    return chain


def Next(vectorSeq, q = 0):
    if (vectorSeq == [1 for i in range(len(vectorSeq))]):
        return [0 for i in range(len(vectorSeq))]
    for i in range(len(vectorSeq) - 1, -1, -1):
        if (vectorSeq[i] == 1):
            vectorSeq[i] = q + i + 1
            continue
        vectorSeq[i] -= 1
        break
    return vectorSeq


def Split(n):
    for m in range(int(math.log(n, 2)), int(math.log(n, 2)) + Nu(n) + 1):
        q = m // 2
        if (q == 0):
            q = 1
        rSeq = [i for i in range(1, q + 1)]
        rhoSeq = [q + i + 1 for i in range(m - q)]
        while (rSeq != [0 for i in range(q)]):
            chain = MakeChain(rSeq + rhoSeq)
            a_min = chain[q] + (m - q)
            a_max = chain[q] * 2 ** (m - q)
            if (n < a_min):
                rSeq = Next(rSeq)
                continue
            elif (n > a_max):
                rSeq = Next(rSeq)
                continue
            if(chain[-1] == n):
                return chain
            while (rhoSeq != [0 for i in range(m - q)]):
                chain = MakeChain(rSeq + rhoSeq)
                if (n == chain[-1]):
                    return chain
                rhoSeq = Next(rhoSeq, q)
            rSeq = Next(rSeq)
            rhoSeq = [q + i + 1 for i in range(m - q)]
    return []


table_res = []
array = [505, 513, 555, 632]
for n in array:
    start = time.time()
    temp = Split(n)
    end = time.time()
    print('n =', n, ', l(n) =', len(temp))
    print('Потрачено времени:', end - start, ' секунд')
    print('Звездная цепочка:', temp)
    print()
    row = [n, len(temp), end - start, temp]
    table_res.append(row)
    
    
    
headers = ['n', 'l(n)', 'Время (сек)', 'Звездная цепочка'] 
cell_text = []
for i in table_res:
    cell_text.append([data for data in i])

table = plt.table(cell_text, rowLoc='right', colLabels=headers, loc='center')
table.set_fontsize(15)
table.auto_set_column_width(col=list(range(3)))
table.scale(5, 4) 

ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)

plt.draw()  
```

### 4. Проверка следствия теоремы Брауэра 

```python
import pylab as plt
import math

def functionLn(n):
    l_n = [len(Split(i)) for i in range(1, n)]
    x = [i for i in range(1, n)]
    x_ln = [math.log(i, math.e) for i in range(1, n)]
    x_ln2 = [math.log(i, math.e)**2 for i in range(1, n)]
    y_i = [l_n[i-1] * x_ln[i-1] for i in range(1, n)]
    f = squaresMethod(n, y_i, x_ln, x_ln2, l_n)
    return x, f, x_ln, l_n

def squaresMethod(n, y_i, x_ln, x_ln2, l_n):
    b = (n*sum(y_i) - sum(l_n)*sum(x_ln))/(n*sum(x_ln2) - sum(x_ln)*sum(x_ln))
    a = (sum(l_n) - b * sum(x_ln))/(n-1)
    f = [a + b*x_ln[i-1] for i in range(1, n)]
    return f
 
    
n = 201
x, y, x_ln, l_n = functionLn(n)


plt.figure(figsize=(10, 10))
plt.scatter(x, l_n, color = 'red', marker = '.')
plt.plot(x, y, color = 'purple')
plt.show()
```

#### Выводы

Смотря на график, можно сделать вывод, что значения длин аддитивных цепочек апроксимируется логарифмической кривой (подтверждается следствием теоремы Брауэра). Также значение длины аддитивной цепочки и функции логарифма эквивалентны на бесконечности, из этого можно сделать вывод, что на бесконечности длина аддитивной цепочки будет расти очень медленно.
