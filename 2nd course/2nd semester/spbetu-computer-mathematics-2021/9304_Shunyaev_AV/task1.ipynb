## Основные теоретические положения

### Бинарный метод
Данный метод позволяет быстро возводить число в степень $n \in N$. Для использования этого метода степень записывается в двоичном виде. После этого перебираются цифры полученного двоичного числа слева направо. Если цифра 0, то число возводится в квадрат. Если цифра 1, то число возводиться в квадрат и умножается на исходное число. Старший бит в двоичной записи числа отбрасывается. Для подсчета количество операций можно воспользоваться формулой: $\lambda(n)+\nu(n)-1$, где $\lambda(n)=[ log_2 n ],  \nu(n)$ - количество едениц в двоичной записи числа.  

### Метод множителей
Метод множителей основывается на том, что если степень $n \in N$ - составное число, то n можно представить как $n = k * m$. Тогда, если мы можем вычислить $x^k$ за i операций и $x^m$ за j операций, то $(x^k)^m$ можно вычеслить за $i + j$ операций. При этом k должно быть наименьшим простым множителем n.

### Определение аддитивной цепочки
Аддитивной цепочкой для $n \in N$ называется последовательность $1=a_1, a_2, a_3, \dots, a_m=n$, где $a_i=a_j+a_k$, $k\leq j<i$ $\forall i=1..m$. То есть каждый элемент последовательности равен сумме каких-либо двух предыдущих элементов.

Минимальная длина аддитивной цепочки обозначается как $l(n) = m$, где m - длина цепочки.

### Теорема Брауэра
При некотором $k < log_2 log_2 n$ справедливо следующее неравество: 
$$l(n)<(1+k^{-1}) \cdot \lceil log_2 n \rceil + 2^{k-1}-k+2$$

При $k=\lambda(\lambda(n))-2\lambda(\lambda(\lambda(n)))$, можно получить следующие следствия:

Следствие 1: $$\lim \limits_{n \to \infty} \frac{l(n)}{\lambda(n)}=1$$

Следствие 2: Одна из наилучших верхних оценкок для длины аддитивной цепочки имеет следующий вид: 

$$\lambda(n)(1+\frac{1}{\lambda(\lambda(n))}+\frac{o(\lambda(\lambda(\lambda(n))))}{\lambda(\lambda(n))^2})$$

### Приближенный алгоритм Яо
Алгоритм вычисляет аддитивную цепочку для числа n. Задается число k>=2. Затем число n представляется в виде:
$$n = \displaystyle\sum_{i=0}^{j} a_i*2^{ik} , a_j\neq0$$

Аддитивная цепочка Яо состоит из 2 аддитивных цепочек, первая цепочка равна $Y_k(n):1,2,4,\ ...\, 2^{\lambda (n)}$, вторая аддитивная цепочка составляется на основе первой. $d(z)$ - сумму всех разрядов, цифра которых равняется z

Затем получаем $$z*d(z) ,
\forall z \in \{1,2,3,...,2^k-1\},\  d(z)\neq0$$

В итоге: $n=\displaystyle\sum_{z=1}^{2^{k-1}}z*d(z)$

Вычислительная сложность алгоритма Яо:
$$lb(n) + \frac {(1+o(1))*lb(n)}{lb(lb(n))}$$

### Алгоритм дробления вектора индексов
Вектор индексов - последовательность натуральных числе длиной m-1, по которой можно построить аддитивную звездную цепочку длиной m. Каждый элемент в векторе - это индекс элемента звездной цепочки, который нужно прибавить к i-му элементу, чтобы получить i+1 элемент.\ Данный алгоритм нужен для того, чтобы найти для заданного числа $n\in N$ минимальную звездную цепочку. Алгоритм сначала с помощью внешнего цикла перебирает все длины m вектора индексов, где $\underline{l}(n) \le m \le \overline{l}(n)$. $\underline{l}(n) = \lceil lbn \rceil$, $\overline{l}(n) = \lambda(n) + \nu(n)-1$. Далее происходит деление вектора индексов на 2 части: $\{r_i\}_{i=1}^q$ и $\{\rho\}_{j=q+1}^m$, где $1\le q\le m-1$. Далее с помощью внутреннего цикла происходит перебор первой части вектора, в нем считается максимальное $a_{max}$ и минимальное значение $a_{min}$ итогового числа. Если $a_m \in [a_{min};a_{max}]$ , то с помощью второго внутреннего цикла происходит перебор всех 2 частей вектора, в ином случае алгоритм переходит на следующий вектор. Если $a_m = n$, то алгоритм завершает свою работу.

### Гипотеза Шольца-Брауэра
Гипотеза Шольца-Брауэра для звездной цепочки заключается в $l^*(2^n-1)\le l^*(n)+n-1$, где $l^*(n)$ - длина минимальной аддитивной цепочки для числа n.

#### Бинарный метод

```python
def find_nu(n):
    if n == 1:
        return 1
    
    return (n%2) + find_nu(n//2)

def binary_method(x, n):
    binary_n = bin(n)[3:]
    
    defult_x = x
    
    for bit in binary_n:
        if bit == '1':
            x *= x
            x *= defult_x
        else:
            x *= x
            
    operation_counter = floor(log(n,2)) + find_nu(n) - 1
    return x, operation_counter

print("Бинарный метод:")
x = int(input("  Введите x: "))
n = int(input("  Введите n: "))

res_x, operations = binary_method(x,n)

print(f"  {x} в степени {n} = {res_x}")
print(f"  Количество операций = {operations}")
```


#### Метод множителей

```python
def multiplier_method(x,n):
    multipliers = []
    primes = list(factor(n))
    oper_counter = 0
    current_x = x
    
    for prime in primes:
        for i in range(prime[1]):
            multipliers.append(prime[0])
    
    for multiplier in multipliers:
        if multiplier < 14:
            x, b_operations = binary_method(x, multiplier)
            oper_counter += b_operations
            current_x = x
        else:
            x, m_operations = multiplier_method(x, multiplier - 1)
            x *= current_x
            oper_counter += m_operations + 1
            current_x = x
            
    return x, oper_counter

print("Метод множителей:")
x = int(input("  Введите x: "))
n = int(input("  Введите n: "))

res_x, operations = multiplier_method(x,n)

print(f"  {x} в степени {n} = {res_x}")
print(f"  Количество операций = {operations}")
```

### Алгоритм Яо
Реализация алгоритма Яо

```python
def zdz(dz, z):
    if(z == 1):
        return [dz]
    elif(z == 2):
        return [dz, dz+dz]
    
    orig_dz = dz
    chain = [dz]
    factor_z = list(factor(z))
    prime_flag = False
    
    if(factor_z[0][0] == z):
        z = z - 1
        prime_flag = True
    
    bin_z = bin(int(z))[3:]
    for i in bin_z:
        if i == '0':
            dz += dz
        else:
            dz += dz
            chain.append(dz)
            dz+=orig_dz
        chain.append(dz)
    if(prime_flag):
        chain.append(dz+orig_dz)
    return chain

def yao(n, k):
    chain = []
    i = 1
    while i < n:
        chain.append(i)
        i = i*2
    
    digits = n.digits(2**k)
    
    j = 1
    result = 0
    while j < (2**k):
        dz = 0
        for index in range(len(digits)):
            if digits[index] == j:
                dz += (2**(k*index))
                if dz not in chain:
                    chain.append(dz)
                
        if dz != 0:
            temp_zdz = zdz(dz, j)
            for x in temp_zdz:
                if x not in chain:
                    chain.append(x)
                    
            result += temp_zdz[-1]
            if result not in chain:
                chain.append(result)
        j += 1
    return chain
    
    
n = Integer(input("Введите n: "))
k = Integer(input("Введите k: "))

chain = yao(n,k)

print("Длина цепочки: ", len(chain))
print("Аддитивная цепочка: ", chain)
```

### Алгоритм дробления вектора индексов
Релизован алгоритм дробления вектора индексов для нахождения минимальной звёздной цепочки для числа n.

```python
import math
import time

def star_chain(vector):
    chain = [1]
    for i in vector:
        chain.append(chain[-1] + chain[i-1])
    return list(chain)

def next_vector(vec,length):
    add = 0
    if length != len(vec):
        add = length - len(vec)
    if vec == [i + add for i in range(1,len(vec)+1)]:
        return [0 for i in range(len(vec))]
    
    for i in range(len(vec)-1,-1,-1):
        if vec[i] < (i + 1 + add):
            vec[i] += 1
            return vec
        vec[i]=1

def split_vector(n):
    if n == 1:
        return [1]
    for m in range(int(math.log(n,2)), int(math.log(n,2))+ int(bin(n).count('1') + 1)):
        q = m // 2
        if q == 0:
            q = 1
            
        r=[1 for i in range(q)]
        rho = [1 for i in range(q+1,m)]

        while r!=[0 for i in range(q)]:
            chain = star_chain(r + rho)
            a_min = chain[q]+m-q
            a_max = chain[q]*2**(m-q)
            
            if(chain[-1]==n):
                return chain
            if(n < a_min and n > a_max):
                r = next_vector(r,len(r))
                continue
                
            while rho != [0 for i in range(q+1,m)]:
                chain = star_chain(r + rho)
                if chain[-1] == n:
                    return chain
                rho = next_vector(rho,len(rho)+len(r))
            r = next_vector(r,len(r))
            rho = [1 for i in range(q+1,m)]
```

### Проверка следствия теоремы Брауэра


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import leastsq

x = [i for i in range(1,201)]
y = [len(split_vector(i)) for i in range(1,201)]

def residuals(coeffs, y, t):
    return  y - np.log(t)*coeffs[0]

figure, axes = plt.subplots(figsize=(10,10))
p, flag = leastsq(residuals, 1, args=(y, x))
plt.plot(x,y,'o',x, np.log(x)*p[0])
plt.show()
```

#### Вывод
Данный график подтверждает следствие из теоремы Брауэра: 
$$\lim\limits_{n \to \infty}\frac{l(n)}{\lambda(n)}=1$$
