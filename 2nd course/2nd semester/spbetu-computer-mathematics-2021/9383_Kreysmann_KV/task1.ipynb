### Основные теоретические положения


#### Аддитивная цепочка 
Аддитивная цепочка — последовательность натуральных чисел, начинающаяся с единицы, в которой каждый последующий элемент является суммой каких-то двух предшествующих элементов (в том числе, возможно использование одного и того же предшествующего элемента — удвоение).

---

#### Типы шагов

$\forall i \in [1,m]$, где $m$ - длина аддитивной цепочки, $a_{i}= a_{j} + a_{k}$, 0$\leq$k$\leq$j$\leq$i.

1) $i$-ый шаг - удвоение, если $j=k=i-1$

2) $i$-ый шаг - звездный, если $j=i-1$, $k \in[0,i-1]$.

3) $i$-ый шаг - малый, если $\lambda(a_{i}) = \lambda(a_{i-1}). (\lambda(n) = \lfloor lb ( n ) \rfloor )$ .

---

#### Звездная цепочка
Аддитивная цепочка называется звездной, если каждый ее шаг является звездным. 

l(n) - длина кратчайшей аддитивной цепочки, l*(n) - длина звездной аддитивной цепочки.

Гипотеза Шольца-Брауэра: $$l^*(2^n - 1) \leq n + l^*(n) -1$$

---


#### Бинарный метод
   Для построения аддитивной цепочки для числа n бинарным методом необходимо перевести n в двоичную систему счисления, убрать первую единицу, затем пройтись по оставшимся нулям и единицам слева направо: если попадается 0, то производится удвоение; если попдается 1, то производится удвоение, а затем прибавляется единица. 
   
   Длина бинарной аддитивной цепочки: $$l(n) = \lambda(n)+\nu(n) -1$$
   
---
    
#### Метод множителей
   Для построения аддитивной цепочки для числа n методом множителей необходимо разложить число в виде $n=m*k$, где m - наименьший простой множитель k. Затем получить аддитивные цепочки для m и k: $$1,a_{1},...,a_{i}=m;1,b_{1},...,b_{j}=k$$
   
   После чего получим аддитивную цепочку для n: $$1,a_{1},...,a_{i-1},m,m*b_{1},...,m*b_{j-1},m*k=n$$
   
   Если для вычисления аддитивный цепочек для m и k использовать бинарный метод, то получим:$$l(n) = \lambda(m)+\lambda(k)+\nu(m)+\nu(k) -2$$
   
---
   
#### Теорема Брауэра
   При $k< lb(lb(n))$ справедливо $$l(n) < (1+k^{-1}) *  \lceil lb(n) \rceil + 2^{k-1}-k+2$$ 
   
   Cледствия:
   $$1)\lim\limits_{n \to \infty}\frac{l(n)}{\lambda(n)}=1$$ 
   2) Наилучшая верхняя оценка для $l(n)$ :
   $$\lambda(n) *(1+\frac{1}{\lambda(\lambda(n))}+\frac{o(\lambda(\lambda(\lambda(n))))}{(\lambda(\lambda(n)))^2})$$ 

---
   
#### Алгоритм Яо

Алгоритм вычисляет аддитивную цепочку для числа n. Задается число k>=2. Затем число n представляется в виде:
$$n = \displaystyle\sum_{i=0}^{j} a_i*2^{ik} , a_j\neq0$$

Вводится функция $d(z)$ - сумма чисел $2^{ik}$ $\forall i : a_i=z$.

Аддитивная цепочка Яо начинается с $$1,2,4,8,...,2^{jk}$$

Затем получаем $$z*d(z) ,
\forall z \in \{1,2,3,...,2^k-1\},\  d(z)\neq0$$

В итоге: $$n=\displaystyle\sum_{z=1}^{2^{k-1}}z*d(z)$$

Вычислительная сложность алгоритма Яо:
$$lb(n) + \frac {(1+o(1))*lb(n)}{lb(lb(n))}$$

---

#### Алгоритм дробления вектора индексов
Пусть дано число n, алгоритм дробления вектора индексов ищет минимальную звёздную цепочку $a_1,a_2,...,a_m=n$. 

*Опеределение*: Пусть дана звёздная цепочка $a_1,a_2,...,a_m$, $l^*(n)=m$. Вектором индексов называется последовательность $\{r_1,r_2,...,r_{m-1}\}$, такая что $r_i = \{x: 1\leq x \leq i\}$ и $a_i=a_{i-1} + a_{r_{i-1}}$, $(2 \leq i \leq m-1)$.

Максимальной звёздной цепочке $1,2,4,...,2^m$ соответствует вектор индексов $R=\{ 1,2,...,m-1 \}$.

Минимальной звёздной цепочке $1,2,3,4,...,m$ соответствует вектор индексов $R=\{ 1,1,1,...,1 \}$.

*Определение*: Пусть даны два вектора индексов $R = \{ r_i \}_{i=1}^{m-1}$ и $\widetilde R \{ \widetilde r_i \}_{i=1}^{m-1}$.

$R=\widetilde R$ ,если $ r_2=\widetilde r_2, r_3=\widetilde r_3, ... , r_{i-1} = \widetilde r_{i-1}, r_i>\widetilde r_i $.

Данный алгоритм является переборным. Изначально вектор индексов разбивается на два вектора : $\{ r_i \}_{i=1}^q$ - фиксированный и $\{  \rho_j \}_{j=q+1}^m$ - меняющийся.

Наибольшее значение a_m достигается при $$\{ r_i \}_{i=1}^q \cup \{q+1,...,m-1\};\ \ \ a_{max} = a_{q+1} * 2^{m-q}$$

Наименьшее значение a_m достигается при $$\{ r_i \}_{i=1}^q \cup \{1,...,1\};\ \ \     a_{min} = a_{q+1} +m -q$$

Для фиксированного вектора вычисляются $a_{min}$  и  $a_{max}$. Если $ n < a_{min}$ или $ n>a_{max}$, то алгоритм переходит к следующему фиксированному набору. Если $n \geq a_{min}$ и $ n \leq a_{max} $, то перебирается меняющийся вектор. Если $a_m=n$, то алгоритм завершает свою работу.  Если все наборы вектора индексов длины $m$ исчерпаны, то увеличивается m на 1 и вышеперечисленные действия повторяются.

$m$ перебирается от $\underline{l}(n)$ до $\overline{l}(n)$, где $\underline{l}(n) = \lceil lb(n) \rceil $, а $\overline{l}(n) = \lambda(n) + \nu(n) -1$.

```python
def binary1(n,power):
    result = n
    count = 0
    kuk = bin(power)[3:]
    for i in kuk:
        if i == '0':
            result*=result
            count+=1
        else:
            result*=result
            result*=n
            count+=2
    return result,count           
```

```python
def min_prime1(n):
    if n % 2==0:
        return 2
    for i in range(3,int(n/2)+1,2):
        if(n%i==0):
            return i
    return n
        

def multiplier1(n,power):
    i=min_prime1(power)
    prime =False
    if(i==power):
        i=2
        power-=1
        prime=True
    j=power/i
    result,count1 = binary1(n,i)
    result,count2 = binary1(result,j)
    count=count1+count2
    if(prime):
        count+=1
        result*=n
    return result,count
    
```

#### Алгоритм дробления вектора индексов
Релизован алгоритм дробления вектора индексов для нахождения минимальной звёздной цепочки для числа n.



```python
import math
import time
def nextVector(n,length):
    add = 0
    if length != len(n):
        add = length - len(n)
    if n==[i + add for i in range(1,len(n)+1)]:
        return [0 for i in range(len(n))]
    for i in range(len(n)-1,-1,-1):
        if n[i]<i+1 + add:
            n[i]+=1
            return n
        n[i]=1
def getChain(vec1,vec2):
    chain = [1]
    for i in vec1:
        chain.append(chain[-1]+chain[i-1])
    for j in vec2:
        chain.append(chain[-1]+chain[j-1])
    return chain
def splitting(n):
    if(n==1):
        return [1]
    for m in range(int(log(n,2)),math.ceil(log(n,2))+bin(n).count('1')+1):
        q=m//2
        if q==0:
            q=1
        r=[1 for i in range(q)]
        rho = [1 for i in range(q+1,m)]
        while r!=[0 for i in range(q)]:
            chain = getChain(r,rho)
            a_min,a_max = chain[q]+m-q, chain[q]*2**(m-q)
            if(chain[-1]==n):
                return chain
            if(n<a_min and n>a_max):
                r= nextVector(r,len(r))
                continue
            while rho!=[0 for i in range(q+1,m)]:
                chain = getChain(r,rho)
                if chain[-1] == n:
                    return chain
                rho = nextVector(rho,len(rho)+len(r))
            r = nextVector(r,len(r))
            rho = [1 for i in range(q+1,m)]
#n = 711
#start = time.time()
#chain = splitting(n)
#sec = time.time()-start
#print("Звёздная цепочка для числа",n,":{",end=' ')
#for i in chain:
#    print(i,end=' ')
#print("}")
#print("Длина:",len(chain))
#print("Время работы алгоритма: ",sec,"секунд")

```

#### Проверка следствия теоремы Брауэра

$$\lim\limits_{n \to \infty}\frac{l(n)}{\lambda(n)}=1$$

```python
import pylab as plt
max_n=200
arrLen = [len(splitting(i)) for i in range(1,max_n+1)]
```

```python
N = len(arrLen)
sumYi = sum(arrLen)
sumLnXi = sum([log(i) for i in range(1,max_n+1)])
sumLnXi2 = sum([log(i)**2 for i in range(1,max_n+1)])
sumYiLnxi = sum(arrLen[i]*log(i+2) for i in range(N))
A = (sumYi*sumLnXi2 - sumYiLnxi*sumLnXi)/(N*sumLnXi2 - sumLnXi**2)
B = (N*sumYiLnxi - sumYi*sumLnXi)/(N*sumLnXi2 - sumLnXi**2)

plt.figure(figsize=(10, 10))
plt.scatter([i for i in range(1,max_n+1)], arrLen, color='g', marker='.')
plt.plot([i for i in range(1,max_n+1)], [A+B*log(i) for i in range(1,max_n+1)], color = 'purple')
plt.show()
 
```

#### Вывод

Из графика видно, что $l(n)$ аппроксимируется логарифмической кривой, то есть при увеличении $n$ - $l(n)$ будет расти как логарифмическая функция, что подтверждает следствие из теоремы Брауэра: $$\lim\limits_{n \to \infty}\frac{l(n)}{\lambda(n)}=1$$

Приходим к заключению, что чем больше число $n$, тем, при его увеличении, будет меньше расти длина минимальной аддитивной цепочки.

#### Проверка гипотезы Шольца–Брауэра
$$l^*(2^n - 1) \leq n + l^*(n) -1$$


```python
arr1 = [i + len(splitting(i)) - 1 for i in range(2,11)]
arr2 = [len(splitting(2**i-1)) for i in range(2,11)]
```

```python
for i in range(len(arr1)):
    print("n+l*(n)-1 = ",arr1[i])
    print("l*(2^n-1) = ",arr2[i])
    print()
    
```
