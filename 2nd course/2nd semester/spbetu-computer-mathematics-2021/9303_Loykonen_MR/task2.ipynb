## Основные теоретические сведения
#### Определение 1
Пусть $G$ - группа с нейтральным элементом $e$, $X$ - множество.
$G$ действует на $X$, если задана операция $G \times X \rightarrow X$ такая, что $\forall x \in X$, $ \forall g, h \in G$ выполняются свойства:

1) $ex = x$

2) $g(hx) = (gh)x$

#### Определение 2
Орбитой $x \in X$ под действием группы $G$ называется множество $Gx=\{gx | g \in G\}$.

Длина орбиты равна количеству элементов орбиты.

#### Определение 3
Неподвижными точками элемента $g \in G$ называются такие $x \in X$, для которых выполняется: $gx = x$.

$X^{g} - $ множество неподвижных точек элемента $g$.

#### Лемма Бернсайда
Количество орбит действия группы $G$ на $X$ равно $N = \frac{1}{\#G} \sum_{g \in G}{\#X^{g}}$

#### Определение 4
Пусть $I$ - множество, $C$ - множество цветов.

Раскраской множества $I$ называется функция из $I$ в $C$.

$C^I$ - множество всех раскрасок.

$\#C^{I}=(\#C)^{\#I}$ - количество всех раскрасок.

#### Определение 5
Пусть $G$ - группа, действующая на множестве $I$.

Тогда формула $gf(i)=f(g^{-1}i)$, задает действие $G$ на $C^{I}$.

$g \in G$, $i \in I$, $f \in C^{I}$

#### Определение 6
Пусть $g \in G$, $f \in C^{I}$. Эквивалентны следующие условия:

1) $gf=f$

2) $f(i) = f(g^{n}i)$, $\forall i \in I$, $\forall n \in \mathbb{Z}$

3) Если $i$, $j \in I$ лежат в одном цикле в циклической записи перестановки $\phi(g)$, то $f(i)=f(j)$.

#### Определение 7
Количество раскрасок из $C^{I}$, которые сохраняют элемент $g \in G$, равно $\#C^{I}=(\#C)^{C(\phi(g))}$.

#### Лемма 1
Пусть $\tau$ является циклом длины $k$, а НОД$(k, n) = d$.

Тогда перестановка $\tau^n$ является произведением $d$ циклов длины $\frac{k}{d}$.

## Реализация графического отображения раскраски

```python
def generateNumber():
    number = []
    for i in range (0, 77):
        number.append(randint(0, 4))
    return number

def printBeads(transposition, number):
    listColors = ['red', 'green', 'blue', 'purple', 'orange']
    colors = {'red': [], 'green': [], 'blue': [], 'purple': [], 'orange': []}
    g = graphs.CycleGraph(77)
    g.relabel({x: transposition[x][0] for x in range(0, 77)})
    for i in range(0, len(number)):
        colors[listColors[number[i]]].append(i)
    g = g.plot(vertex_size=200, vertex_colors=colors)
    g.show(figsize=[10, 8])
    return

def rotate(transposition):
    tmp = transposition[0][0]
    for i in range(0, 76):
        transposition[i][0] = transposition[i+1][0]
    transposition[76][0] = tmp
    return transposition

def symmetry(transposition):
    transposition[1:] = transposition[76:0:-1]
    return transposition

transposition = []
for i in range(0, 77):
    transposition.append([i, i])
number = generateNumber()
printBeads(transposition, number)
```
