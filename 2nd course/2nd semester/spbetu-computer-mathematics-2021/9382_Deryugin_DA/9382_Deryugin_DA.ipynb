# Основные теоретические положения

## Аддитивная цепочка

Аддитивная цепочка — последовательность натуральных чисел, начинающаяся с единицы, в которой каждый последующий элемент является суммой каких-то двух предшествующих элементов. То есть $$1 = a_0, a_1, ..., a_m = n;$$ 
$$a_i = a_j + a_k,   k<=j<i, \forall i > 1$$

## Бинарный метод возведения числ в степень

Для того, чтобы возвести число $x$ в некую степень $k$, нужно представить степень в двоичной системе счисления, удалить из его записи первый бит и каждый бит, равный 1 заменить на $ку$, а каждый нулевой бит заменить на $к$, где $к$ - умножение числа на себя, а $у$ - умножение на основание. Например:$$k = 10_{10} = 1010_2.$$ Тогда, путем замены битов на $к$ и $у$ и удаления первого бит получим: $ккук$  или же такую последовательность действий: $$x, x^2, x^2x^2=x^4, x^4x=x^5, x^5x^5=x^{10}.$$ При таком способе количество операций равно: $\lambda(k) + \nu(k) - 1$, где $\lambda(k)$ - уменьшенная на единицу длина бинарной записи числа $к$, а $\nu(k)$ - число единиц в бинарной записи числа $к$ 

## Метод множителей

Мы хотим посчитать $x^k$. Возьмем $i$ - наименьший простой множитель числа $k$ и $j$, такое что $ij=k$. Тогда для того, чтобы найти $x^k$ нужно: посчитать $x^i = y$ бинарным методом, а затем посчитать $y^j$ также бинарным методом. Конечный результат будет нашим числом. Если $k$ - простое число, то сначала ищем $x^{k-1}$, а затем на $x$.

## Алгоритм Яо

Алгоритм Яо помогает найти приближенно минимальную аддитивную цепочку для числа $n$. Сначала выбирается число $k>=2,$ а  $n$ раскладывается на сумму вида:$\sum_{i=0}^jC_i2^{ik}$, где $С_i!=0.$ После этого мы ищем все $d(z)$ - суммы чисел $2^{ik}$ для всех $i$, таких что $C_i=z.$ Теперь нам остается найти $d(z): \forall z \in \left\{1,2,..., 2^k-1 \right\},$ а также $zd(z):\forall z:n=\sum_{z=1}^{2^k-1}zd(z).$ Конечая аддитивная цепочка для $n$ будет иметь следующий вид: $$1,2,3,..., 2^{jk}, d(1), 2d(1), ..., zd(z), d(2), 2d(2),..., zd(z), n.$$

## Алгоритм дробления вектора индексов

Данный алгоритм находит одну из минимальных аддитивных цепочек для числа n. 
Введем несколько понятий. $$$$ Звездный шаг: $a_i = a_{i-1} + a_k  \forall k \in [0,i - 1]$ $$$$
Вектор индексов: $ \left\{r_i \right\}_{i=1}^{q} \cup \left\{\rho_j \right\}_{j={q+1}}^{m}$, где первое множество - это фиксированная часть вектора индексов, а вторая часть - изменяемая.$$$$
$a_m = a_{max}:\left\{r_i \right\}_{i=1}^{q} \cup \left\{q+1, q+2...m \right\} = a_{q+1}*2^{m-q}$ - максимальный вектор индексов $$$$
$a_m = a_{min}:\left\{r_i \right\}_{i=1}^{q} \cup \left\{1, 1...1 \right\} = a_{q+1} - m - q$ - минимальный вектор индексков, где количество единиц равно m-q. $$$$
Минимальная аддитивная цепочка будет равна m: $l_{min}(n) \leq m \leq l_{max}(n)$, где $l_{min}(n) =  \lfloor log_2(n)\rfloor$, а $l_{max}(n) = \lambda(n) + \nu(n) - 1$ $$$$
Сам алгоритм:
1. Во внешнем цикле перебираем все $m \in [l_{min}(n), l_{max}(n)]$
 Выбираем $q=m/2$, где $1 \leq q \leq m-1$
2. Во внутреннем цикле перебираем все вариации фиксированной части вектора индексов и поибавляем к ней максимальную и минимальную изменяющиеся части. Находим $a_{min}, a_{max}$. Если последний элемент построенной звездной цепочки оказался n, то алгоритм завершен. Если же n не лежит между $a_{min} и a_{max}$, то переходим к следующему фиксированному набору. Если n лежит в этом интервале, тогда переходим к пункту 3.
3. Перебираем все варианты изменяющей части вектора индексов.

Следствие теоремы Брауэра гласит:$\lim\limits_{n \to \infty} \frac{l(n)}{\lambda(n)} = 1$

## Гипотеза Шольца–Брауэра

Гипотез Шольца-Брауэра гласит: $l^*(2^n - 1)\leq l^*(n) + n - 1$

### 1. Алгоритмы нахождения степени бинарным методом и методом множителей

```python
def count_of_iterations_binary_method(power):
    iterations = 0
    sequence = [1]
    #convert to binary without lead 1
    binary_power = bin(power)[3:]
    
    for letter in binary_power:
        if letter == '1':
            iterations+= 2
            sequence.append(sequence[-1]*2)
            sequence.append(1+sequence[-1])
        else:
            iterations+= 1
            sequence.append(sequence[-1]*2)
    if sequence[-1] != power:
        sequence.append(power)
    
    return sequence
```

```python
def factor(power):
     for number in range (2, power+1):
        if power % number == 0:
            return number
    
def count_of_iterations_multipliers_method(power):
    factor_of_power = factor(power)
    if factor_of_power == power:
        return count_of_iterations_binary_method(power - 1) + [power]
    first_part = count_of_iterations_binary_method(factor_of_power) 
    
    second_part = count_of_iterations_binary_method(int(power/factor_of_power))[1::]
    second_part = [i * first_part[-1] for i in second_part]
    
    return first_part + second_part
```

### 2. Реализация алгоритма Яо

```python
#conver number from 10 to 2^k base
def convert_base(n, base):
    newNum = ''
    digits = 0
    while n > 0:
        newNum = ',' + str(n % base) + newNum
        n //= base
        digits+=1
    return newNum[1:]

#sort dictinary in alphabetic order
def dict_sort(dictinary):
    new_dict = {}
    tmp_list = list(dictinary.keys())
    tmp_list.sort();
    for a in tmp_list:
        new_dict[a] = dictinary[a]
    return new_dict


def find_d_z(n, base):
    result = {}
    iteration = 1
    length = len(n)
    for digit in n:
        if int(digit) != 0:
            num = 2**(base*(length - iteration))
            if digit in result.keys():
                num+= int(result[digit][-1])
                tmp_array = result[digit] + [num]
            else:
                tmp_array = [num]
                
            result.update({digit:tmp_array})
        iteration+= 1
        
    return result
        
        
def print_sequence(array_of_k_z, lambda_n):
    chain = [1]
    sequence_number = 1
    while sequence_number <= 2**lambda_n:
        sequence_number*=2
        chain.append(sequence_number)
    
    for k_z in array_of_k_z:
        for item in k_z[0]:
            if k_z[1] * item in chain:
                continue
            chain.append(k_z[1] * item)
   
    return chain
   
    
def Yao(n, k):
    base = 2**k
    lambda_n = len(bin(n)) - 4
    # перевод в систему счисления 2^k
    converted_num = convert_base(n, base)
    
    # поиcк d(z)
    d_z = find_d_z(converted_num.split(','), k)
    

    array_of_k_z = []
    extra_arr = []
    for key, values in d_z.items():        
        if isinstance(values, list):
            for item in values:
                 extra_arr.append(item)
        array_of_k_z.append([count_of_iterations_binary_method(int(key)), values[-1]])
    

    chain = print_sequence(array_of_k_z, lambda_n)
    if n not in chain:
        chain.append(n)
    for item in extra_arr:
        if item not in chain:
            chain.append(item)
    return [chain, len(chain)]
    
        
```

### 3. Алгоритм дробления веторов

```python
import time


def is_trivial_vector(vector):
    for a in vector:
        if a != 1:
            return False
    return True

def mutable_chain(chain_kind, length, q = -1):
    if chain_kind == 'min':
        return [1 for x in range(1, length)]
    if chain_kind == 'max':
        return [x + q for x in  range(1, length + 1)]

def star_chain(chain):
    star_chain_var = [1]
    [star_chain_var.append(star_chain_var[-1] + star_chain_var[y - 1]) for y in chain]
    return star_chain_var
    
    
def change_vector(vector, q = 0):
    index = len(vector) - 1
    for item in reversed(vector):
        if item > 1:
            vector[index] = item - 1
            for x in range(index + 1, len(vector)):
                vector[x] = x + q + 1
            return vector
        index-=1
    return vector

    
def vector_splitting_algorithm(n):
    if n == 1:
        return [n, 1, [1], 0]
    if n == 3:
        return [n, 3, [1, 2, 3], 0]
    if n == 4:
        return [n, 3, [1, 2, 4], 0]
    if n == 5: 
        return [n , 4, [1, 2, 3, 4], 0]
    if n == 6:
        return [n, 4, [1,2, 4, 6], 0]
    start_time = time.process_time()
    #find range of length
    lambda_min = math.floor(math.log2(n))
    lambda_max = math.floor(math.log2(n)) +  bin(n).count('1') - 1
    index_vector = []
    
    for m in range(lambda_min, lambda_max + 1):
        index_vector = []
        q = m // 2
        if q == 0:
            q = 1        
        #fill fixed chain
       
        [index_vector.append(x) for x in range(1, q + 1)]
        while not is_trivial_vector(index_vector):
            mutable_chain_vector =  mutable_chain('max', m - q, q)
           
            min_star_chain = star_chain(index_vector + mutable_chain('min', m - q))
            max_star_chain = star_chain(index_vector + mutable_chain('max', m - q, q))
            a_min = min_star_chain[-1]
            a_max = max_star_chain[-1]
            if a_min == n:
                return [n, len(min_star_chain), min_star_chain,  time.process_time() - start_time]
            if a_max == n:
                return [n, len(max_star_chain), max_star_chain,  time.process_time() - start_time]
            
            if n > a_min and n < a_max:
                while not is_trivial_vector(mutable_chain_vector):
                    mutable_chain_vector = change_vector(mutable_chain_vector, q)
                    tmp_star_chain  = star_chain(index_vector + mutable_chain_vector)
                    if tmp_star_chain[-1] == n:
                        return [n, len(tmp_star_chain), tmp_star_chain, time.process_time() - start_time]  
            index_vector = change_vector(index_vector)
    last_chain = star_chain(index_vector)
    return [n, len(last_chain), last_chain ,  time.process_time() - start_time]
            
            
        
        
    
```
