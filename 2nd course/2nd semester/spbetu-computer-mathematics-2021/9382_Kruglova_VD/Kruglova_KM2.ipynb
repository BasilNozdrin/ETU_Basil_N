## Основные теоретические положения

### Определение 1
Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая для $\forall x \in X$ и $\forall g,h \in G$ следующими двумя свойствами: 

1) $ex = x$

2) $g(hx) = (gh)x$

где $gx$ - это образ пары $(g,x)$.

###  Определение 2
Пусть $X$ - множество. Множество всех биективных функций с $X \to X$ с операцией композиции называется симметрической группой на множестве $X$ и обозначается как $S_{X}$

### Определение 3
Орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в данной орбите называется длиной орбиты. Любые две орбиты либо не пересекаются, либо совпадают. Таким образом, множество $X$ разбивается в дизъюнктивное объединение орбит.

### Определение 4
Неподвижными точками элемента $g \in G$ называются те $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента g обозначается через $X^{g}$.

### Теорема Бёрнсайда
Количество орбит действия группы $G$ на множестве $X$ равно:

$\frac{1}{|G|}\sum_{g \in G} |X^{g}|$

### Лемма 1
Предположим, что $\sigma$ является циклом длины $k$, а $GCD(k,n) = d$. Тогда перестановка $\sigma^{n}$ является произведением d циклов длины $\frac{k}{d}$.

### Определение 5
Пусть $I$ - произвольное множество, а $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$. Множество всех раскрасок обозначается через $C^{I}$. 

### Лемма 2
Пусть $G$ группа, действующая на множестве $I$. Тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G$, $i \in I$, а $f \in C^{I}$ задает действие $G$ на $C^{I}$.

### Лемма 3
Для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:

1)$gf = f$

2)$f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$

3)Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$ (в терминологии задачи о бусах, это означает, что все места из одного цикла покрашены в одинаковый цвет)

### Следствие 1
Количество раскрасок из $C^{I}$, которые сохраняет данный элемент $g \in G$ равно $|C^{c(\varphi(g))}|$, где $c(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$ (считая циклы длины 1).

### Задание
Сколькими способами можно составить бусы из 145 бусинок 5 различных цветов (с точностью до поворотов и осевых симметрий)?

### Решение
Имеем упорядоченный набор $\{ x_{1}, x_{2}, x_{3}, ... , x_{145} \}$, где $x_{i} \in \{ a, b, c, d, e \}$ - различные цвета. 

Существует всего $5^{145}$ таких раскрасок, но не все из них соответствуют различным бусам.

Есть две операции после применения которых будут получаться одинаковые бусы - поворот и симметричное отображение.

Пример поворота ($\tau$): $\{1, 2, 3, 4, 5, 6\} \rightarrow \{ 6, 1, 2, 3, 4, 5 \}$.

Примет симметричного отображения ($\sigma$): $\{ 1, 2, 3, 4, 5, 6 \} \rightarrow \{ 6, 5, 4, 3, 2, 1\}$.

Таким образом на множестве всех раскрасок $X$ действует группа $G$ поворотов и осевых симметрий, задано отображение из $G$ в симметрическую группу $S_{145}$.

Для перестановки $\{ 1, 2, 3, ... ,145 \}$ после применения поворота получается перестановка $\{ 145, 1, 2, 3, ..., 144 \}$. Цикл соответствующий этой перестановке будет $(1,2,3,4,...,145)$ и его длина равна $k = 145$. Для нахождения количества $d$ циклов в перестановках $\tau^{n}$ воспользуемся Леммой 2, т.е. будем применять формулу $GCD(k,n) = d$. 

Множители числа $145$ - это простые числа $5$ и $29$. Тогда для всех $n$ не кратных $5$ и $29$, $d$ будет принимать значение 1, для кратных $5$ - $5$, для кратных $29$ - $29$.

Теперь рассмотрим перестановку после операции $\sigma$. Этой операции будет соответствовать перестановка $(1, 145)(2, 144)(3, 143)...(73)$, количество циклов равно 73. Также после применения перестановки вида $\sigma \tau^{n}$ количество циклов останется быть равным 73. Так происходит из-за "симметричности" перестановки.

Раскраски задают одинаковые бусы тогда и только тогда, когда они лежат в одной орбите под действием $G$. Таким образом, надо вычислить количество орбит действия $G$ на $X$.

Для того, чтобы воспользоваться леммой Бёрнсайда, необходимо выяснить, сколько элементов множества $X$ оставляет на месте каждый элемент группы $G$. Воспользуемся Следствием 1 (что для подсчёта таких элементов множества $X$ нужно лишь знать количество независимых циклов). Количества независимых циклов:

$|X^{e}| = 5^{145}$

$|X^{\tau^{i}}| = 5^{1}$,  $i \in \{1..144\} / \{ i | (i\%5 = 0) || (i\%29 = 0)\}$ (112 значений)

$|X^{\tau^{5}}| = |X^{\tau^{10}}| = |X^{\tau^{15}}| = ... = |X^{\tau^{140}}| = 5^{5}$ (28 значений)

$|X^{\tau^{29}}| = |X^{\tau^{58}}| = |X^{\tau^{87}}| = |X^{\tau^{116}}| = 5^{29}$ (4 значения)

$|X^{\sigma}| = |X^{\sigma \tau}| = |X^{\sigma \tau^{2}}| = ... = |X^{\sigma \tau^{144}}| = 5^{73}$ (145 значений)

$|G| = 290$


По лемме Бёрнсайда количество различных бус равно:

$\frac{1}{290}(5^{145} + 112*5^{1} + 28*5^{5} + 4*5^{29} + 145*5^{73}) = 773130187213692177061368045953057175878765209311710511044209542616630023044778903753593050200363989$

### Трансформация и графическое отображение бусин

```python
bead_color = ["#DDDD65", "#2CC7ED", "#BB2CED", "#2CED7E", "#8A7D92"]
initial_necklace = [[i,bead_color[random.randint(0,4)]] for i in range(1,146)]
```

```python
def do_rotate(necklace):
    return [necklace[-1]] + necklace[:-1]

def do_symmetry(necklace):
    return necklace[::-1]
```

```python
def print_necklace(necklace):
    G = graphs.CycleGraph(145)
    recolor_dict = {'#DDDD65' : [], '#2CC7ED' : [], '#BB2CED' : [], '#2CED7E' : [], '#8A7D92' : []}
    relabel_dict = {i:necklace[i][0] for i in range(0,145)}
    for item in initial_necklace:
        recolor_dict[item[1]].append(item[0])
    G.relabel(relabel_dict)
    P = G.graphplot(vertex_colors=recolor_dict, vertex_size=220)
    P.show(figsize=[9,9])
```

### Стартовые бусы:

```python
print_necklace(initial_necklace)
```

Бусина выше в свою очередь получилась путем симметричного отображения начальной бусины. Как видно, действительно, эта буса отражена относительно главной оси.

Также можно скомбинировать несколько операций, давайте применим $\sigma \tau^{4}$ к начальной бусе:

```python
necklace_3 = do_rotate(do_rotate(do_rotate(do_rotate(do_symmetry(initial_necklace)))))
print_necklace(necklace_3)
```

Как можно заметить, буса выше и в правду сначала отобразилась относительно главной оси, а потом повернулась влево на 4 позиции.

```python
import random
def transform_necklace(necklace, times):
    logs = []
    for idx in range(0, times):
        r = random.randint(1,2)
        if r == 1:
            necklace = do_rotate(necklace)
            logs.append([necklace,"rotation done"])
        else:
            necklace = do_symmetry(necklace)
            logs.append([necklace,"symmetry done"])
    return logs
```

Теперь давайте применим 20 случайных операций и рассмотрим пару состояний бус.

```python
states = transform_necklace(initial_necklace, 20)
state_logs = [item[1] for item in states]
print(state_logs)
```
