## Основные теоретические положения
**1) Бинарный метод**

Бинарный метод предполагает возведение любого числа в степень натурального числа N за O(log n) произведений. Метод имеет следующий алгоритм: число переводится в двоичный вид, и, начиная с старшей степени двойки, число возводится в квадрат. Далее, после возведения в квадрат, переходим на степень двойки, на единицу меньше чем ранее рассмотренную, и выполняется аналогичное действие(число возводится в квадрат). Если рассматриваемый бит равен единице, то полученное после возведения в квадрат число умножается на исходное. Выполняются эти действия до тех пор, пока не придем к 0-вой степени двойки.  

**2) Аддитивные цепочки**

Аддитивная цепочка $n\in N$ предполагает последовательность натуральных чисел до $a_m$. При этом эти числа обладают следующими свойствами: 

 $1=a_1, a_2, a_3, \dots, a_m=n$, где $a_i=a_j+a_k$, $k<j<i$ $\forall i=1..m$


**3) Метод множителей**

Метод сомножителей - это метод, предполагающий быстрое возведение в степень, при этом степень n раскладывается на 2 множителя. Метод имеет следующий алгоритм: Изначально степень раскладывается на 2 множителя, выбирается меньший множитель, после чего число возводится в степень, равную этому множителю, потом возводится в степень, равную оставшемуся множителю. В ситуации, если n будет простым числом, то исходное число сначала будет возводить число в степень n-1, после чего будет домножаться на исходное число

**4) Алгоритм дробления вектора индексов**

Алгоритм дробления индексов предполагает нахождение минимальной звёздной цепочки, такой, что $l*(n)=m-1:a_m=n$. Для описания работы алгоритма необходимо ввести понятие вектора индексов.

Вектор индексов - это последовательность натуральных чисел, каждый из которых длиной $m-1$, при этом, в этом векторе каждый элемент является индексом элемента звёздной цепочки, который нужно прибавить к i-тому элементу, чтобы получить $i+1$ элемент.

Алгоритм состоит в следующем: изначально разобьем вектор индекса на 2: фиксированный и изменяющийся(при этом разбиение происходит посередине). После чего, берём фиксированную часть и вычисляем $a_{max}$ и $a_{min}$. Если n меньше $a_{min}$, то переходим к следующему фиксированному вектору в сторону уменьшения или в случае, если n больше a..max, то наоборот, переходим к другому фиксированному вектору в сторону увеличения. Но если n попало в промежуток между $a_{min}$ до $a_{max}$ включительно, то производим перебор по меняющемуся вектору.Алгоритм завершает работу в случае, если $a_m$ будет равен $n$.


**5) Алгоритм Брауэра**

Брауэрская цепочка при $n \in N$ задаётся рекуррентной формулой: $ q = \left\lfloor \dfrac{n}{2^{k}} \right\rfloor $, если $k \geq 2$. Алгоритм позволяет проводить поиск $ q_{s} $ и $ r_{i} $ в виде $ n $: $ n = 2^{k} \left( 2^{k} \left( \dots \left( 2 q_{s} + r_{s} \right) \dots \right) + r_{2} \right) + r_{1} $. Соответственно, $ B_{k} \left( n \right) = 1 $, $ 2 $, $ \dots $, $ 2^{k} - 1 $, $ 2 q_{s} $, $ 4q_{s} $, $ \dots $, $ 2^{k} q_{s} $, $ 2^{k} q_{s} + r_{s} $, $ \dots $, $ n $, где $ B_{k} \left( n \right)$ - это брауэрская цепочка.


**6) Алгоритм Яо**

Алгоритм яо предполагает вычисление аддитивной цепочки для натурального n, при этом он не гарантирует минимальность аддитивной цепочки.

Для описания алгоритма следует внести понятие функции $d(z)$:


$d(z)=\displaystyle\sum_{i=0} a_i \cdot 2^{ik}$

В начале алгоритма аддитивная цепочка Яо представляет из себя последовательность степеней двойки:
${1,2,4,...,2^k}$

После чего находится следующее произведение:
$z*d(z)$, откуда $n=\displaystyle\sum_{z=1}^{2^{k-1}} z \cdot d(z)$

**7) Теорема Брауэра**

Учитывая, что $k<log_2 log_2 n$ следует что $l(n)<(1+k^{-1}) \cdot \lceil log_2 n \rceil + 2^{k-1}-k+2$
Из этой теоремы следует следствие:
При  $k=\lambda(\lambda(n))-2\lambda(\lambda(\lambda(n)))$ следует $\lim \limits_{n \to \infty} \frac{l(n)}{\lambda(n)}=1$


**8) Гипотеза Шольца-Брауэра**

$l^*(2^n-1) \geq l^*(n)+n-1$

**1) Бинарный метода и метод множителей:**

Были реализованы 2 функции: Binary_fnc и Multipl_fnc. Binary_fnc используется для выполнения бинарного метода, а Multipl_fnc - для метода множителей соответственно, код функций представлен ниже.


```python
import matplotlib.pyplot as plt
import pandas as pd

iter = 0


def Binary_fnc(value, checked):
    global iter
    if checked == 0:
        iter += 1
        return 1
    elif checked == 1:
        return value

    cheked_kr_2 = checked % 2
    iter += 1
    if cheked_kr_2 == 1:
        return Binary_fnc(value, checked - 1) * value
    elif cheked_kr_2 == 0:
        return Binary_fnc(value, checked / 2)^2


def Multipl_fnc(value, checked):
    factors_n = []
    d = 2
    while 1:
        if checked % d == 0:
            factors_n.append(d)
            factors_n.append(checked // d)
            break
        else:
            d += 1
    ans = value
    for i_loc in factors_n:
        ans = Binary_fnc(ans, i_loc)
    return ans


headers = ['Степень N', 'Количество операций в ММ', 'Количество операций в БМ']
L = []
line = []

i=100
while i<100000:
    line = []
    Multipl_fnc(2, i)
    line = [str(i), iter]
    iter = 0
    Binary_fnc(2, i)
    line.append(iter)
    iter = 0
    L.append(line)
    i*=10

fig, ax = plt.subplots()
ax.set_axis_off()
table = ax.table(
    cellText=L,
    colLabels=headers,
    cellLoc='center',
    loc='upper left',
)

ax.set_title('Сравнение бинарного метода и метода множителей', fontsize=15, fontweight='bold', loc='left')
table.set_fontsize(15)
table.scale(2, 3)
plt.show()
```

Было проведено сравнение бинарного метода и метода множителей при n=100,1000,10000.Из чего был сделан вывод, что метод множителей в некоторых случаях работает быстрее, но в основном, оба метода обрабатываются с одинаковой скоростью.

**3) Алгоритм дробления вектора индексов**

Ниже представлены функции, реализующие алгоритм дробления вектора индексов, который необходим для построения звёздной цепочки, код функций представлен ниже:



```python

import matplotlib.pyplot as plt
import time
import math

def splitting_i_v(n):
    for m in range(math.ceil(math.log(n, 2)) - 1, math.ceil(math.log(n, 2)) + bin(n)[2:].count("1") - 1 + 1):
        q = m // 2 if m != 1 else 1
        fix = [i for i in range(1, q + 1)]
        while 1:
            changed = [i for i in range(q + 1, m + 1)]
            time_i_v= fix + changed
            chn_s = [1]
            il=0
            while il<len(time_i_v):
                chn_s.append(chn_s[-1] + chn_s[il - 1])
                il+=1
            min_a = chn_s[q] + m - q
            max_a = chn_s[q] * 2 ** (m - q)
            if chn_s[-1] == n:
                return chn_s
            if min_a <= n <= max_a:
                while sum(changed) != len(changed):
                    i = len(changed) - 1
                    while i >= 0 and changed[i] == 1:
                        i -= 1
                    if i != -1:
                        changed[i] -= 1
                        i += 1
                    while i < len(changed):
                        changed[i] = q + i + 1
                        i += 1
                    time_i_v= fix + changed
                    chn_s = [1]
                    for il in time_i_v:
                        chn_s.append(chn_s[-1] + chn_s[il - 1])
                    if chn_s[-1] == n:
                        return chn_s
            if sum(fix) == len(fix):
                break
            i = len(fix) - 1
            while i >= 0 and fix[i] == 1:
                i -= 1
            if i != -1:
                fix[i] -= 1
                i += 1
            while i < len(fix):
                fix[i] = i + 1
                i += 1
    #raise




headers = ['Значение n', 'Длина', 'Время', 'Звёздная цепочка']
L = []
line = []
n=500
k=1
while n<510:
    n+=5
    line = []
    start_time = time.time()
    chain = splitting_i_v(n)
    line = [str(n),str(len(chain)), str(time.time() - start_time), str(chain)]
    L.append(line)

fig, ax = plt.subplots()
ax.set_axis_off()
table = ax.table(
    cellText=L,
    colLabels=headers,
    cellLoc='center',
    loc='upper left',
)

ax.set_title('Алгоритм дробления вектора индексов', fontsize=15, fontweight='bold', loc='left')
table.set_fontsize(15)
table.scale(5, 3)
table.auto_set_column_width([0,1,2,3])
plt.show()
```

Тестирование проводилось для n=500 и n=510. Как и предполагалось, алгоритм находит минимальную аддитивную цепочку, в отличии от алгоритма Брауэра, который, находит приблизительную минимальную аддитивную цепочку.
