### Определение №2

### Определение №3
Длина орбиты - количество элементов в орбите.

### Определение №4

Неподвижные точки элемента $g\in \mathbb{ G}$ называют те  $x \in \mathbb{X}$, для которых выполняется  $gx=x$.

### Лемма Бёрнсайда
Количество орбит действия группы $\mathbb{G}$ на $\mathbb{X}$ равно

$$N = \frac{1}{\left | \mathbb{G} \right |}\sum_{g \in \mathbb{G}}\left | \mathbb{X}^g \right |$$

### Определение №5

Пусть $I$  - произвольное множество,  $C$  - множество цветов. Раскраской множества  $I$  называется функция из  $I$  в  $C$ .

### Определение №6

Пусть $\mathbb{G}$ - группа, действующая на множестве $I$ . Тогда формула

$$gf(i)=f(g^{-1}i),$$

где $g \in \mathbb{G}$, $i \in  I$, а $f \in C ^{ I},$ задает действие $g$ на $ C^{ I}$.

### Теорема

Для элемента $g \in \mathbb{G}$ и функции  $f \in C^I$  следующие условия эквивалетны:

1.  $gf=f$ 


2.  $f(i)=f(g^ni)$,  для всех  $i \in I$ и  $n \in \mathbb{Z}$


3. Если  $i, j \in I$ лежат в одном цикле в циклической записи перестановки  $\phi(g)$, то $f(i)=f(j)$.

### Следствие

Количество раскрасок из $C^I$, которые сохраняет данный элемент $g \in \mathbb G$ равно ${\left | C \right|}^{c(\phi(g))}$.

Круг разбит на 105 секторов, каждый из которых покрашен в один из 3 цветов. Сколькими способами можно составить такую мозаику (с точностью до поворотов круга)?


## Выполнение работы

**Аналитическое решение задачи.** 

Дано множество цветов $ C = \{a, b, c\} $ и множество секторов $\mathbb{X} = \{x_1, x_2, ..., x_{105}\} $ в круге. Количество всех раскрасок равно $ 3 ^ {105} $, но некоторые из них переходят друг в друга под влиянием вращения круга.

Пусть $ \mathbb{G}$ - группа преобразований множества $\mathbb{X}$, $ e $ - тождественное преобразование, запишем его в виде перестановки: 

$ e = \begin{pmatrix}
       1 & 2 & 3 &  \ldots & 104 & 105 \\
       1 & 2 & 3 & \ldots & 104 & 105 \\
       \end{pmatrix}
       = (1)(2)(3)...(104)(105)
$

$\sigma$ - поворота на один сектор. Мощность множества поворотов равно количеству секторов $|\mathbb G| = 105 $. 

Приведём несколько примеров перестановок, получившихся под действием группы $\mathbb G$ на $\mathbb X$:



$ \sigma = \begin{pmatrix}
       1 & 2 & 3 & \ldots & 104 & 105 \\
       2 & 3 & 4 & \ldots & 105 & 1 \\
       \end{pmatrix}
       = \begin{pmatrix}
       1 & 2 & 3 & \ldots & 104 & 105 \\
       \end{pmatrix}
$


$ \sigma^2 = \begin{pmatrix}
       1 & 2 & 3 & \ldots & 104 & 105 \\
       3 & 4 & 5 & \ldots & 1 & 2 \\
       \end{pmatrix}
       = \begin{pmatrix}
       1 & 3 & 5 & \ldots & 103 & 105 & 2 & 4 & \ldots & 102 & 104 \\
       \end{pmatrix}
$

$ \sigma^5 = \begin{pmatrix}
       1 & 2 & 3 & \ldots  & 103 & 104 & 105 \\
       6 & 7 & 8 & \ldots  & 3 & 4 & 5\\
       \end{pmatrix}
       = \begin{pmatrix}
       1 & 6 & \ldots & 100\\ 
       \end{pmatrix}
       \begin{pmatrix}
       2 & 7 & \ldots & 102\\ 
       \end{pmatrix}
       \begin{pmatrix}
       3 & 8 & \ldots & 103\\ 
       \end{pmatrix}
       \begin{pmatrix}
       4 & 9 & \ldots & 104\\ 
       \end{pmatrix}
       \begin{pmatrix}
       5 & 10 & \ldots & 105\\ 
       \end{pmatrix}
$

Воспользуемся леммой Бернсайда, чтобы найти количество различных раскрасок, но до этого нужно выяснить количество независимых циклов во всех перестановках, полученных после применения $ \sigma ^ n $. Для этого воспользуемся леммой 2. Число 105 делится на 3, 5, 7, 15, 21, и 35. 

|  $d$ |  $X$ | $n$  |
|:-:|:-|:-:|
|3|$ \{3, 6, 9, 12, 18, 24, 27, ..., 99, 102\} $| 24|
|5|$ \{5, 10, 20, 25, 40, 50, 55, 65, 80, 85, 95, 100\} $| 12 |
|7|$ \{7, 14, 28, 49, 56, 77, 91, 98\} $| 8 |
|15|$ \{15, 30, 45, 60, 75, 90\} $|6|
|21|$ \{21, 42, 63, 84\} $|4|
|35|$ \{35, 70\} $|2|


В тождественном преобразовании количество независимых циклов равно $ 105 - 24 - 12 - 8 - 6 - 4 - 2 = 49 $. Тогда по лемме Бёрнсайда количество различных раскрасок будет равно

$ \frac{1}{105}(3^{105}+24*3^3+12*3^5+8*3^7+6*3^{15}+4*3^{21}+2*3^{35}+49*3) $


```python
import matplotlib.pyplot as plt

def draw(colors, n = 0):
    n = n % 105
    sizes = [360/105 for _ in range(105)]
    labels = [str(i+1) for i in range(105)]
    plt.figure(figsize = (18,18))
    plt.pie(sizes,
            labels = labels, 
            colors = colors, 
            labeldistance = 1.01, 
            radius = 1,
            startangle = 360/105*n+90,
            wedgeprops = {"linewidth": 1, 'width':1, "edgecolor":"black"}
            ) 
    plt.show()
```

```python
from random import randint 
clr =['tomato', 'yellow', 'teal']
colors = [clr[randint(0, 2)] for i in range(105)]
```

```python
print("начальное состояние круга")
draw(colors)
```

```python
n = 15
print(f"поворот круга на {n} секторов")
draw(colors, n)
```

```python
n = 12345
print(f"поворот круга на {n} ({n} mod 105 = {n%105}) секторов")
draw(colors, n)
```
