## Основные теоретические положения

### 1. Бинарный метод
Такой метод возведения в степень, который позволяет возводить любое число в натуральую степень. Основной идеей метода является разбиение искомого числа ${x^n}$ на произведение ${x}$ в некоторых степенях. Результат считается по следующим правилам:
1. Слева направо просматривается бинарная запись числа ${n}$ без первого бита, так как он всегда равен 1.
2. Если текущий бит - это ${1}$, то текущее число возводится в квадрат, после чего еще раз умножается на ${x}$.
3. Если текущий бит - это ${0}$, то текущее число возводится в квадрат.

При использовании данного метода количество операций умножения вычисляется по формуле: ${\lambda (n) + \nu (n) - 1}$, где ${\lambda (n) = \left \lfloor lb(n) \right \rfloor}$, а ${\nu(n)}$ - вес Хэмминга для бинарного представления числа ${n}$, иначе говоря - число ненулевых битов в двоичной записи числа 

### 2. Метод множителей

Этот метод также служит для возведения числа в натуральную степень и базируется на вышеописанном (бинарном) методе. Дано число ${x}$ и натуральная степень ${n}$, в которую это число предстоит возвести.
1. Находим такое число ${i}$, которое является наименьшим простым множителем числа ${n}$.
2. Вычисляем число ${j}$, такое что: ${i*j=n}$.
3. Возодим число ${x}$ в полученную ранее степень ${i}$, затем полученное число возводим в степень ${j}$.
4. Если число ${n}$ - простое, то сначала вычисляется ${x^{-1}}$, которое после домножается на ${x}$.

Допустим, возможно посчитать ${x^i}$ за ${m}$ операций, а ${y^j}$ за ${k}$ операций, где ${y = x^i}$. Тогда утверждается, что ${x^{i \cdot j}}$ можно посчитать за ${(m + k)}$ операций.

### 3. Аддитивная цепочка
Аддитивная цепочка — это такая последовательность натуральных чисел ${a_0, a_1, a_2, ..., a_m = n}$, в которой первый элемент всегда равен ${1}$, а каждый последующий элемент является суммой каких-то двух предшествующих элементов. То есть:
- ${a_0 = 1}$ 
- ${\forall i > 0, a_i = a_j + a_k , j,k < i}$ - шаг аддитивной цепочки.

Виды шагов:
- ${i}$-шаг - удвоение, если ${j = k = i - 1}$
- ${i}$-шаг - звездный, если ${j = i - 1}$, ${k = \overline{0 .. (i-1)}}$
- ${i}$-шаг - малый, если ${\lambda(a_i) = \lambda(a_{i-1})}$

### 4. Алгоритм Яо
Алгоритм Яо вычисляет аддитивную цепочку для натурального числа ${n}$. Прежде всего выбирается число ${k}$ и, в случае, когда ${n,k \leq 2}$, ${n}$ представляется в следующем виде:
${n = \sum_{i=0}^{j} a_i  2^{ik}, a_j \ne 0}$.

Определим функцию ${d(z)}$, как сумму чисел ${2^{i*k}, \forall i: a_{i} = z}$.

В начале работы аддитивная цепочка Яо будет иметь следующий вид: 1, 2, 4, 8, ..., ${2^k}$.

После чего находим ${z*d(z);  \forall z \in \{1,2,3,...2^{k-1}\}, d(z) \neq 0}$.

Тогда число ${n}$ примет следующий вид: ${n = \sum_{z=1}^{2^{k-1}} z*d(z)}$.

Также следует помнить, что алгоритм Яо не гарантирует, что полученная аддитивная цепочка является минимальной.

### 5. Звездная цепочка
Звездная цепочка - это аддитивная цепочка, включающая только звездные шаги. Подробнее данный шаг был описан в пункте 3.

### 6. Алгоритм дробления вектора индексов
Данный алгоритм используется для нахождения минимальной звездной цепочки для числа ${n}$.

Пусть у нас есть звездная цепочка ${a_i}$, длины ${l(n) = m-1}$.

Тогда для каждой такой звездной цепочки можно задать вектор индексов вида ${\left \{ r_1, r_2, ..., r_{m-1} \right \} : r_i = \left\{x: 1 \leq x \leq i\right\}}$. Тогда можно записать наш шаг в виде ${a_i = a_{i-1} + a_{r_{i-1}}}$.

Для поиска минимальной звездной цепочки осуществим перебор. Если осуществлять его "в лоб", то число операций будет равно ${(m-1)!}$, что недопустимо много. Для уменьшения числа операций разделим наш вектор индексов на две части:  фиксированную и остаток, который может меняться: ${\left\{r_i\right\}_{i=1}^q \cup \left\{\rho_i\right\}_{j=q+1}^m}$, где ${\rho_{j} = \left\{ x : 1 \leq x \leq j \right\}}$.

Заметим также, что наибольший ${a_m}$ будет получаться в цепочке ${\left\{r_i\right\}_{i=1}^q \cup \left\{q+1, q+2, ..., m\right\}}$, а наименьший ${\left\{r_i\right\}_{i=1}^q \cup \left\{1, 1,..., 1\right\}}$. Соответственно, получаем формулы:
1. ${a_{min} = a_{q+1} + m - q}$
2. ${a_{max} = a_{q+1} \cdot 2^{m-q}}$

После рассчета вышеупомянутых значений проводится проверка: 
- Если ${n<a_{min}}$ или ${n>a_{max}}$, то разбиение меняется.
- Если ${n \in [a_{min}, a_{max}]}$, то происходит перебор меняющегося вектора.
- Если ${n=a_m}$, то задача считается решенной и алгоритм прекращает работу.
- Если ${\{\rho_j\}}$ исчерпан, то происходит переход к следующему набору ${\{r_i\}}$.
- Если и ${\{\rho_j\}}$ и ${\{r_i\}}$ исчерпаны, то увеличиваем длину аддитивной цепочки на 1.

### 7. Теорема Брауэра
При некотором ${k < lb(lb(n))}$ справедливо следующее неравенство ${l(n) < (1 + k^{-1})\left \lceil lb(n) \right \rceil + 2^{k-1} - k + 2}$.

#### 7.1 Следствие 1 из теоремы Брауэра
Если ${k = \lambda(\lambda(n)) - 2\lambda(\lambda(\lambda(n)))}$, то ${\lim_{n \to \infty} \frac{l(n)}{\lambda(n)} = 1}$

#### 7.2 Следствие 2 из теоремы Брауэра
Наименьшая верхняя оценка: ${\lambda(n)(1 + \frac{1}{\lambda(\lambda(n))} + \frac{\mathcal{O}(\lambda(\lambda(\lambda(n))))}{{\lambda(\lambda(n))}^2}}$

### 1. Вычисление степеней бинарным методом и методом множителей

#### 1.1. Релизация бинарного метода
Ниже представлена программная реализация бинарного метода на языке программирования Python. Программа производит возведение в степень, а также считывает количество произведенных операций.

```python
def binaryMethod(x, n): #функция принимает на вход число x и натуральную степень n
    tmpx = x
    tmpbin = bin(n)[3:]
    stepcounter = 0
    for i in tmpbin:
        if (i == '1'):
            tmpx = tmpx**2
            tmpx = x*tmpx
            stepcounter += 2
        elif (i == '0'):
            tmpx = tmpx**2
            stepcounter += 1
    return (tmpx, stepcounter)
```

#### 1.2. Релизация метода множителей
Ниже представлена программная реализация метода множителей на языке программирования Python. Программа производит возведение в степень, а также считывает количество произведенных операций.

```python
import math
def minPrime(n):
    for i in range(2, int(math.sqrt(n))+1):
        if (n%i == 0):
            return i
    return 1

def coefMethod(x, n):
    minprime = minPrime(n)
    tmpx, stepcounter1 = binaryMethod(x, minprime)
    result, stepcounter2 = binaryMethod(tmpx, n//minprime)
    return (result, stepcounter1+stepcounter2)
```

#### 1.3. Сравнение бинарного метода и метода множителей
Сравнение произведено для значений степени больше 30: 33, 500, 1921. Результаты представлены в виде сравнительной таблицы.

```python
x = 77
n = [33, 34, 55]
for i in n:
    tmpx1, stepcounter1 = binaryMethod(x,i) 
    tmpx2, stepcounter2 = coefMethod(x,i)
    print("Степень n:"+str(i))
    print("Количество операций бинарным методом = "+str(stepcounter1))
    print("Количество операций методом множителей = "+str(stepcounter2))
```

### 2. Алгоритм Яо

#### 2.1. Реализация
Ниже представлена программная реализация меода Яо.

```python
import math
def dfunction(z, d):
    tmpd = d
    result = [tmpd]
    binaryz = bin(z)[3:]
    for i in binaryz:
        if (i == '0'):
            tmpd += tmpd
            result.append(tmpd)
        elif (i == '1'):
            tmpd += tmpd
            result.append(tmpd)
            tmpd += d
            result.append(tmpd)
    return result

def Yao(n, k):
    tmpn = n
    a = []
    while(tmpn != 0):
        a.append(tmpn % 2**k)
        tmpn //= 2**k
    answer = []
    for i in range(int(math.log(n, 2))+1):
        answer.append(2**i)
    result = 0
    for z in range(1, 2**(k)+1):
        d = 0
        for i in range(0, len(a)):
            if (a[i] == z):
                d += 2**(k*i)
                if (d not in answer):
                    answer.append(d)
        if (d == 0):
            continue
        nextd = dfunction(z, d)
        for j in nextd:
            if (j not in answer):
                answer.append(j)
        if (len(nextd) == 0):
            continue
        result += nextd[-1]
        if (result not in answer):
            answer.append(result)
    return answer
```

### 3. Метод дробления вектора индексов

#### 3.1. Реализация
Ниже представлена программная реализация метода дробления вектора индексов

```python
import time
import math

def buildChain(index):
    chain = [1]
    for i in index:
        chain.append(chain[-1] + chain[i-1])
    return chain

def nextVector(vec, q=0):
    if (vec == [1 for i in range(len(vec))]):
        return [0 for i in range(len(vec))]
    for i in range(len(vec) - 1, -1, -1):
        if (vec[i] == 1):
            vec[i] = i+1+q
            continue
        vec[i] -= 1
        break
    return vec

def indexVectorFraction(n):
    for m in range(int(math.log(n, 2)), int(math.log(n, 2)) + bin(n)[3:].count('1') + 1):
        q = m // 2
        if (q == 0):
            q = 1
        r = [i for i in range(1, q+1)]
        rho = [q + i + 1 for i in range(m-q)]
        while (r != [0 for i in range(q)]):
            chain = buildChain(r+rho)
            a_min = chain[q] + (m-q)
            a_max = chain[q] * 2 ** (m-q)
            if ((n < a_min) or (n > a_max)):
                r = nextVector(r)
                continue
            if (chain[-1] == n):
                return chain
            while (rho != [0 for i in range(m-q)]):
                chain = buildChain(r+rho)
                if (chain[-1] == n):
                    return chain
                rho = nextVector(rho, q)
            r = nextVector(r)
            rho = [q + i + 1 for i in range(m - q)]
    return chain

def timeCounter(n):
    start = time.time()
    answer = indexVectorFraction(n)
    end = time.time()
    return (answer, end-start)
```

```python
n = [501, 503]
output = []
for i in n:
    answer, time = timeCounter(i)
    output.append([i, len(answer), time, answer])

print("n        l       time        chain")
for i in range(len(output)):
    print(str(output[i][0])+"       "+str(output[i][1])+"       "+str(output[i][2])+"       "+str(output[i][3]))
```

### 4. Проверка следствия 1 из теоремы Брауэра

#### 4.1. Формулы
Чтобы проверить следствие необходимо ввести ряд формул, которые будут использоваться в логарифмической аппроксимации методом наименьших квадратов.
Мы будем строить функцию ${y=b+a*ln(x)}$, где ${a}$ и ${b}$ - коэфиициенты, вычисляющиеся по следующим формулам:

${a = \frac{n*\sum_{i=1}^{n}(y_i*ln(x_i))-\sum_{i=1}^{n}(y_i)*\sum_{i=1}^{n}(ln(x_i))}{n*\sum_{i=1}^{n}(ln(x_i)^2)-(\sum_{i=1}^{n}(ln(x_i)))^2}}$

${b = \frac{\sum_{i=1}^{n}(y_i)-b*\sum_{i=1}^{n}(ln(x_i))}{n}}$

#### 4.2. Реализация с построением графика

```python
import matplotlib.pyplot as plt
import numpy as np

maxN = 201
dots = [1, 2, 3, 3, 4, 4, 5] + [len(indexVectorFraction(i)) for i in range(8, maxN)]

def aCoef(y):
    n = len(y)
    y = np.array(y)
    x = np.array(range(1, n + 1))
    logx = np.log(x)
    up = n * np.sum(np.multiply(y, logx)) - np.sum(y) * np.sum(logx)
    down = n * np.sum(np.square(logx)) - np.sum(logx) * np.sum(logx)
    return up / down

def bCoef(y, a):
    n = len(y)
    y = np.array(y)
    x = np.array(range(1, n + 1))
    logx = np.log(x)
    up = np.sum(y) - a*np.sum(logx)
    return up / n

def f(a, b, x):
    return a*np.log(x) + b

a = aCoef(dots)
b = bCoef(dots, a)
n = len(dots)
size = 10
plt.figure(figsize=(10, 10))
plt.scatter(range(1, n+1), dots, s=size, color="red")
plt.plot(range(1, n+1), f(a, b, range(1, n + 1)), color='black')
plt.show()
```
