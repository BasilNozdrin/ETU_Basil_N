##Основные теоретические положения

###Определение действия группы на множество.

Пусть $G$ - группа, $e$ - нейтральный элемент, $X$ - множество. Тогда, если задана операция $G \times X->G$, такая что $\forall x \in X; g,h \in G$ выполняются свойства:
1. $cx=x$
2. $g(hx)=(gh)x$

Тогда говорят, что $G$ действует на $x$

###Определение орбиты

Орбита элемента $x$ из $X$ под действием $G$ - это множество $Gx=${$gx|g \in G$}

Количество элементов - длина массива

###Определение неподвижных точек.

Неподвижными точками $g \in G$ называются те $x \in X$, для которых образ меры $gx=x$.

Множество этих точек = $X^g$

###Лемма Бёрнсайда.

Лемма вычисляет количество орбит действия группы на множестве с помощью суммы по всем элементам группы. Как правило, применяется в случаях, когда $\#X>>G$

Количество орбит действия группы $G$ на множество $X$:

N=$1/(\#G)  \sum_{g \in G} \#X^g$

###Определение раскраски

Пусть $I$ - произвольное множество, а $C$ - множество цветов.

Тогда раскраской множества $I$ называется функция $\phi : I->C$

$C^I$ - множество раскрасок

Количество всех раскрасок $\#C^I$= ${\#C}^{\#I}$

###Действие $G$ на $C^I$

Пусть $G$ - группа, действующая на множество I.

Тогда $gf(i)=f(g^{-1}i)$ задает действие $G$ на $C^I$

$g \in G;i \in I;f \in C^I$

###Лемма о количестве раскрасок

Для $g \in G$; $f \in C^I$ эквивалентно:
1. $gf=f$
2. $f(i)=f(g^ni)$ $\forall i \in I, n \in \mathbb{Z}$
3. Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\phi(g)$, то $f(i)=f(j)$

Следствие:

Количество раскрасок из множества $C^I$, которое сохраняет данный элемент $g \in G$ равно $(\#C)^{c(\phi(g))}$, где $c(\omega)$ - количество независимых циклов в записи перестановки $\omega$

Далее, реализуем функцию, выполняющую подсчет циклов в перестановке для конкретного $t$:

```python
def countCycles(arr2):
    numcycles = 0
    checking = 1
    arr3=[]
    arr4=[i+1 for i in range(G_N)]
    """
    print(arr)
    print(arr2)
    """
    while (1):
        if (arr4[checking - 1] == -1):
            """print(arr3)"""
            arr3.clear()
            numcycles += 1
            i = 0
            ch = 0
            while (arr4[i] == -1):
                if (i + 1 == len(arr4)):
                    ch = 1
                    break
                i += 1
            if (ch == 1):
                break
            checking = i + 1
        arr4[checking - 1] = -1
        checking = arr2[checking - 1]
        arr3.append(checking)
    return numcycles
```

###Графически отобразим раскраску по заданному случайно номеру.

Дополним использованный нами ранее код классом enum цветов, возможностью выбирать G_N и G_C,а также подключим библиотеку graphics для создания оконного приложения, в котором отобразим круг с разделенными сегментами своего конкретного цвета.

```python
import math
import random

import graphics
import enum

G_N=385
G_C=5
"""G_C=[2;10]"""

class colorSeg(enum.Enum):
    RED = [255, 0, 0]
    GREEN = [0, 255, 0]
    BLUE = [0, 0, 255]
    YELLOW = [255, 255, 0]
    ORANGE = [255, 128, 0]
    PURPLE = [128, 0, 128]
    PINK = [255, 192, 203]
    CYAN = [0, 255, 255]
    LIGHTGRAY = [211, 211, 211]
    VIOLET = [238, 130, 238]

def countCycles(arr2):
    numcycles = 0
    checking = 1
    arr3=[]
    arr4=[i+1 for i in range(G_N)]
    """
    print(arr)
    print(arr2)
    """
    while (1):
        if (arr4[checking - 1] == -1):
            """print(arr3)"""
            arr3.clear()
            numcycles += 1
            i = 0
            ch = 0
            while (arr4[i] == -1):
                if (i + 1 == len(arr4)):
                    ch = 1
                    break
                i += 1
            if (ch == 1):
                break
            checking = i + 1
        arr4[checking - 1] = -1
        checking = arr2[checking - 1]
        arr3.append(checking)
    return numcycles

arr2=[i+1 for i in range(G_N)]
sum=0
max=0
for j in range(G_N):
    tmp = arr2[-1]
    for i in range(G_N):
        arr2[len(arr2)-i-1]=arr2[len(arr2)-i-2]
    arr2[0]=tmp
    """print(arr2)"""
    k=countCycles(arr2)
    sum+=pow(G_C,k)
print(f"Number of coloring: {(sum//G_N)}")

am=2*math.pi/G_N
l=list(colorSeg)

string=""
for i in range(G_N):
    string+=str(random.randint(0, G_C-1))
print(f"Random numbers: {string}")

G=graphics.GraphWin("Circle",1080,1080)
for i in range(G_N):
    p1 = graphics.Point(540 + 350 * math.sin(am * i), 540 + 350 * math.cos(am * i));
    p2 = graphics.Point(540 + 400 * math.sin(am * i), 540 + 400 * math.cos(am * i));
    p3 = graphics.Point(540 + 400 * math.sin(am * (i+1)), 540 + 400 * math.cos(am * (i+1)));
    p4 = graphics.Point(540 + 350 * math.sin(am * (i+1)), 540 + 350 * math.cos(am * (i+1)));
    pol=graphics.Polygon(p1, p2, p3, p4);
    r=int(string[i])
    pol.setFill(graphics.color_rgb(l[r].value[0],l[r].value[1],l[r].value[2]))
    pol.draw(G)
G.getMouse()
G.close()
```
